```python
numbers = [str(i) for i in range(10)]
print(numbers)
while True:
    try:
        fraction = input("Fraction: ")
        if len(fraction) == 3 and fraction[0] in numbers and fraction[2] in numbers and fraction[1] == "/" and fraction[2] != "0" and int(fraction[0]) =< int(fraction[2]):
            break
    except:
        print("Invalid input, pls try again")
print(fraction)
percent = int((int(fraction[0])/int(fraction[2]))*100)
if percent >= 99:
    print("F")
elif percent <= 1:
    print("E")
else:
    print(str(percent)+"%")
```

```python
while True:
    try:
        fraction = input("Fraction: ")
        if fraction.lower() in ["exit", "quit"]:
            break

        num, denom = fraction.split("/")
        num = int(num)
        print(f"numerator is {num}")
        denom = int(denom)
        print(f"denominator is {denom}")

        if denom == 0 or num > denom:
            raise ValueError
        print("Denominator is zero , raising value error")
        percent = int((num / denom) * 100)
        if percent >= 99:
            print("F")
        elif percent <= 1:
            print("E")
        else:
            print(f"{percent}%")
        break

    except (ValueError, ZeroDivisionError):
        print("Invalid input, pls try again")
```

```python
months = {'january': 1,'february': 2, 'march': 3, 'april': 4, 'may': 5, 'june': 6, 'july': 7,  'august': 8, 'september': 9, 'october': 10, 'november': 11,'december': 12}

while True:
    try:
        user_date = input("Please enter a valid date in month-day-year order format: ")
        ##conditions for 1st format mm/dd/yyyy
        temp_list = user_date.split()
        if user_date.lower() in ["exit", "quit"]:
            break
        elif int(user_date[-4:]) in range(1, 2026) and user_date[-5] == user_date [-8] == "/" and int(user_date[-7:-5]) in range(1, 32) and int(user_date[-10:-8]) in range(1, 13):
            print("You tried numeric inputs")
            formatted_date = user_date[-4:]+"-"+user_date[-10:-8]+"-"+user_date[-7:-5]
            print(formatted_date)
            break 

        elif temp_list[0].lower() in months and int(temp_list[-1]) in range(1, 2025) and int(temp_list[1][:-1]) in range(1,32):
            print("You tried non numeric inputs")
            formatted_date = temp_list[-1]+"-"+str(months[temp_list[0].lower()])+"-"+temp_list[1][:-1]
            print(formatted_date)
            break
    except: 
        print("Invalid inputs, pls try again")
```
