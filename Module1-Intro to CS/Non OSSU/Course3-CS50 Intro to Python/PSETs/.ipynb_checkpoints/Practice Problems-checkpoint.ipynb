```python
numbers = [str(i) for i in range(10)]
print(numbers)
while True:
    try:
        fraction = input("Fraction: ")
        if len(fraction) == 3 and fraction[0] in numbers and fraction[2] in numbers and fraction[1] == "/" and fraction[2] != "0" and int(fraction[0]) =< int(fraction[2]):
            break
    except:
        print("Invalid input, pls try again")
print(fraction)
percent = int((int(fraction[0])/int(fraction[2]))*100)
if percent >= 99:
    print("F")
elif percent <= 1:
    print("E")
else:
    print(str(percent)+"%")
```

```python
while True:
    try:
        fraction = input("Fraction: ")
        if fraction.lower() in ["exit", "quit"]:
            break

        num, denom = fraction.split("/")
        num = int(num)
        print(f"numerator is {num}")
        denom = int(denom)
        print(f"denominator is {denom}")

        if denom == 0 or num > denom:
            raise ValueError
        print("Denominator is zero , raising value error")
        percent = int((num / denom) * 100)
        if percent >= 99:
            print("F")
        elif percent <= 1:
            print("E")
        else:
            print(f"{percent}%")
        break

    except (ValueError, ZeroDivisionError):
        print("Invalid input, pls try again")
```

```python
menu = {
    "Baja Taco": 4.25,
    "Burrito": 7.50,
    "Bowl": 8.50,
    "Nachos": 11.00,
    "Quesadilla": 8.50,
    "Super Burrito": 8.50,
    "Super Quesadilla": 9.50,
    "Taco": 3.00,
    "Tortilla Salad": 8.00}
total = 0
while True:
    try:
        item = input("Input: ")
        print(item)
        if item.lower() in ["exit", "quit"]:
            break
        total += menu[item.title()]
        print("$"+total)

    except:
        print("Invalid inputs pls try again")
```
