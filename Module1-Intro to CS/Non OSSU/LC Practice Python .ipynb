You may assume that each input would have exactly one solution, and you may not use the same element twice.

2 <= nums.length <= 104
-109 <= nums[i] <= 109
-109 <= target <= 109
Only one valid answer exists.
 

Follow-up: Can you come up with an algorithm that is less than O(n2) time complexity?

<h2> Palindrom Number, to be tried through a different way </h2>

Given an integer x, return true if x is a palindrome, and false otherwise.

 

<h3>Example 1:</h3>

Input: x = 121 <br>
Output: true <br>
Explanation: 121 reads as 121 from left to right and from right to left.

<h3>Example 2: </h3>

Input: x = -121 <br>
Output: false <br>
Explanation: From left to right, it reads -121. From right to left, it becomes 121-. Therefore it is not a palindrome.

<h3>Example 3:</h3>

Input: x = 10
Output: false
Explanation: Reads 01 from right to left. Therefore it is not a palindrome.
 

Constraints:

-231 <= x <= 231 - 1

```python
def solution(num):
    num_string = str(num)
    reverse_string = []
    for i in range(len(num_string)-1, -1, -1):
        reverse_string.append(num_string[i])    
    reverse_output = ""
    for i in reverse_string:
        reverse_output += i
    if num_string ==  reverse_output:
        return True
    else:
        return False
    
    
```

<br>

For example, 2 is written as II in Roman numeral, just two ones added together. 12 is written as XII, which is simply X + II. The number 27 is written as XXVII, which is XX + V + II. <br>

Roman numerals are usually written largest to smallest from left to right. However, the numeral for four is not IIII. Instead, the number four is written as IV. Because the one is before the five we subtract it making four. The same principle applies to the number nine, which is written as IX. There are six instances where subtraction is used:<br>

I can be placed before V (5) and X (10) to make 4 and 9. <br>
X can be placed before L (50) and C (100) to make 40 and 90. <br>
C can be placed before D (500) and M (1000) to make 400 and 900.<br>
Given a roman numeral, convert it to an integer.

```python
def solution(roman):
    accepted_list = ['I', 'V', 'X', 'L', 'C', 'D', 'M']
    roman_dict = {
        "I": 1,
        "V": 5,
        "X": 10,
        "L": 50,
        "C": 100,
        "D": 500,
        "M": 1000
    }
                  
    previous_letter = ""
    numeral = 0
    for i in roman:
        if i not in accepted_list:
            return f"Pls enter correct roman letters"
        elif previous_letter == "I" and i in ["V", "X"]:
            numeral +=  roman_dict[i] - 2
        elif previous_letter == "X" and i in ["L", "C"]:
            numeral += roman_dict[i] - 20
        elif previous_letter == "C" and i in ["D", "M"]:
            numeral += roman_dict[i] -200
        else:
            numeral += roman_dict[i]
        previous_letter = i
    return numeral 
```

Can be found at <a href = "https://leetcode.com/problems/find-the-index-of-the-first-occurrence-in-a-string/description/" /a>

```python
def solution(haystack, needle):
    if needle in haystack:
        for i in range(0, len(haystack)):
            if ''.join(haystack[j] for j in range(i, i+len(needle))) == needle:
                return i
    else:
        return -1     
```

<h2> 35. Search Insert Position </h2>

Can be found at <a href = "https://leetcode.com/problems/search-insert-position/description/" /a>

```python
def solution(array, target):
    index_match = 0
    while True:
        if len(array) == 1:
            if array[0] == target:
                return index_match
            elif array[0] > target:
                return index_match
            else:
                return index_match+1
        elif target > array[int(len(array)/2)]:
            change = (int(len(array)/2) + 1) if (len(array) != 2) else int(len(array)/2)
            index_match += change
            array = array[change:]
        elif target < array[int(len(array)/2)]:
            index_match += 0
            array = array[0:int(len(array)/2)]
        else:
            index_match += int(len(array)/2)
            if array[int(len(array)/2)]>target:
                return index_match -1
            elif array[int(len(array)/2)]<target:
                return index_match +1
            else:
                return index_match        
```

Can be found at <a href = "https://leetcode.com/problems/length-of-last-word/description/"> Link </a>

```python
##Need to find which is the last space
def solution(string):
    last_space_index = 0
    for i in range(0, len(string)-1):
        print(f"i is {i} and string[i] is {string[i]} last_space_index is {last_space_index}")
        if string[i] == " " and any(j!=" " for j in string[i+1:]):
            last_space_index = i
    print(f"last_space_index is {last_space_index}")
    final_word = ""
    if last_space_index !=0:
        for i in range(last_space_index+1, len(string)):
            print(f"i is {i} and string[i] is {string[i]}")
            if string[i] != " ":
                final_word += string[i]
        return (len(final_word), final_word)
    else:
        for i in string:
            final_word += i if i != " " else ""
        return len(final_word)
```

<h2> 70. Climbing Stairs </h2>
Link <a href="https://leetcode.com/problems/climbing-stairs/description/"> https://leetcode.com/problems/climbing-stairs/description/ </a>

```python
def solution(num):
    max_giveaway = num//2
    print(f"max_giveway is {max_giveaway}")
    total_count = 0
    for i in range(1, max_giveaway+1): ##lets assume we have 5 spaces [_,_,_,_,_], now we can have 0 and 0 on the last 2 items because we can increase max to 2 the remaining items
        print(f"i is {i}")
        ##num-i factorial
        num_fact = 1
        ##denom factorial
        den_fact = 1
        for j in range(num-i, num-i-i, -1):
            num_fact *= j
        for k in range(i, 0, -1):
            den_fact *= k
        print(f"num_fact is {num_fact} and den_fact is {den_fact}")
        total_count += int(num_fact/den_fact)
        print(f"total count is  {total_count}")
    return total_count + 1
        
```

<h2> 88. Merge Sorted Arrays </h2>

Link <a href = "https://leetcode.com/problems/merge-sorted-array/description/">https://leetcode.com/problems/merge-sorted-array/description/</a>

```python
def solution(nums1, nums2):
    replace_count = 0
    for i in nums2:
        replaced = False
        print(f"value of i is {i}")
        
        # Loop over the currently valid sorted part of nums1
        for j in range(0, len(nums1) - len(nums2) + replace_count):
            print(f"value of i is {i} and value of j is {j} and nums j is {nums1[j]}")
            if i >= nums1[j] and (j + 1 == len(nums1) - len(nums2) + replace_count or i < nums1[j+1]):
                nums1.insert(j+1, i)
                nums1.pop()
                replaced = True
                print(f"nums1 is {nums1} and I will break here and replace count is {replace_count}")
                break

        if not replaced:
            # Insert in correct position in sorted section
            insert_index = 0
            while insert_index < len(nums1) - len(nums2) + replace_count and nums1[insert_index] < i:
                insert_index += 1
            nums1.insert(insert_index, i)
            nums1.pop()
            print(f"nums1 is {nums1} after fallback insert")

        replace_count += 1

    return nums1
```
