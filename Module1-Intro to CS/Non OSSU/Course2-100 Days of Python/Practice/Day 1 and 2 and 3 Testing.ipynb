<h3> adventure game </h3>

```python
print("welcome to trasure island")
direction = input("Chose a direction Left or Right")
if direction == "Right":
    print("game over")
elif direction == "Left":
    step2 = (input("Swim or Wait"))
    if step2 == "Swim":
        print("game over")
    elif step2 == "Wait":
        step3 = input("Which door Red, Blue, Yellow")
        if step3 == "Yellow":
            print("You win")
        elif step3 == "Red" or "Blue":
            print("game over")
        else:
            print("wrong input, try again")
    else:
        print("wrong input, try again")

else:
    print("wrong input, try again")
        
        
    
```
