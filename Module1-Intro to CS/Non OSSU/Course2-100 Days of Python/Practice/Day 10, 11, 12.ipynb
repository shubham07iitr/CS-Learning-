```python
def format_name(f_name, l_name):

    f_name_counter = 0
    l_name_counter = 0
    new_f_name = ""
    new_l_name = ""
    for i in f_name:
        if f_name_counter == 0:
            new_f_name = new_f_name + i.upper()
        else:
            new_f_name = new_f_name + i.lower()
        f_name_counter +=1

    for i in l_name:
        if l_name_counter == 0:
            new_l_name = new_l_name + i.upper()
        else:
            new_l_name = new_l_name + i.lower()
        l_name_counter +=1

    return  new_f_name + " " + new_l_name


format_name("shubham", "gupta")

    
```

```python
def is_leap_year(year):
    """Gives whether a year is leap year or not""" ## this is docstring and will capture documentation of the function
    if year%4 == 0:
        if year%400 == 0:
            return True
        elif year%100 == 0:
            return False
        else:
            return False
    else:
        return False

is_leap_year(2103)
```

<h3> calculator project </h3>

```python
def addition(num1, num2):
    return float(num1) + float(num2)

def subtraction(num1, num2):
    return float(num1) - float(num2)

def multiply(num1, num2):
    return float(num1)*float(num2)

def division(num1, num2):
    return float(num1) / float(num2)


def operation(num1, num2, symbol):
    if symbol == "+":
        return addition(num1, num2)
    elif symbol == "-":
        return subtraction(num1, num2)
    elif symbol == "*":
        return multiply(num1, num2)
    elif symbol == "/":
        return division(num1, num2)
    else:
        return f"Pls chose a correct symbol"



operation(2,3, "+")



```

<h3> No guessing game </h3>

```python
import random

## Asking computer to come up with a number
def random_no():
    computer_choice = random.randint(1,100)
    return computer_choice


def number_game_logic(attempts):
    counter = False
    computer_choice_no = random_no()
    while counter is False:
        for i in range(1,attempts):
            print(f"You have {attempts- i} attempts to guess the number")
            user_choice_no = int(input("Make a Guess"))
            if user_choice_no < computer_choice_no:
                print("Too Low")
                print("Guess again")
            elif user_choice_no > computer_choice_no:
                print("Too high")
                print("Guess again")                
            else:
                print("You got it right, congratulations")
                counter = True
                break
        if counter is False:
            print("Sorry you have run out of guesses, try again")
        else:
            break



def number_game_final(user_choice_difficulty):
    if user_choice_difficulty == "easy":
        number_game_logic(11)
    else:
        number_game_logic(6)



print("Welcome to the Number Guessing Game!\nI'm thinking of a number between 1 and 100.")
user_choice = input("Choose a difficulty, easy or hard")

number_game_final(user_choice)
```
