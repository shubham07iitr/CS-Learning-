<h3>Web Scraping </h3>

- Some websites dont have API in which cases web scraping could be useful
- We look at underlying HTML code to get the info we want
- BeautifulSoup is a python library for pulling data HTML or XML file


```python
##Library for Beautiful Soup
from bs4 import BeautifulSoup
##import lxml to be used if its an XML file


with open("./Day45BS/website.html" , 'r') as f:
    content = f.read()

##This soup object can be used to pull different elements of the html file
soup = BeautifulSoup(content, "html.parser")  ## html parser if html file other wise if xml then use xml parser

print(soup.title)
print(soup.title.name) ##Name of the title
print(soup.title.string) # String inside the title
##print(soup.prettify()) #Entire HTML but indented
print(soup.a) ## First anchor tag
print(soup.li) ## First list element
```

<h3> Finding and selecting particular elements with BS</h3>
**The above methods such as soup.li will give only the first tag how to get all or some of the tags**

```python
##Will give me a list of all anchor tags

all_anchor_tags = soup.find_all(name = "a")

for tag in all_anchor_tags:
    print(tag.getText()) ## Will give me the text which has been hyperlinked
    print(tag.get("href"))
```

```python
## Finding using an id 

heading = soup.find(name = "h1", id = "name") ## will give me 1st element with the required filters
print(heading)
```

**For getting the class attribute we can't use class keyword because its reserved hence we use class_**


```python
section_heading =  soup.find(class_ = "heading")
print(section_heading.getText())
print(section_heading.string)
print(section_heading.name)
print(section_heading.get("class"))
```

**Using CSS selector to select a particular element**

- Useful if there are nesting elements, chaining or descendants where simple find may not give accurate resutls

```python
company_url = soup.select_one("#name")
print(company_url)

headings = soup.select(".heading") ## will give me a list because I am using select and not select_one
print(headings)
```
