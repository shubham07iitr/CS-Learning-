```python
import turtle
from turtle import Turtle, Screen
%gui tk
import pandas as pd
import tkinter.messagebox

## Reading the names of states and their location on the map
data_states =  pd.read_csv("day-25-us-states-game-start/50_states.csv")
data_states.head(1)



screen = turtle.Screen()
screen.title("US States Game")
image  = "day-25-us-states-game-start/blank_states_img.gif"
screen.addshape(image)
turtle.shape(image)

class ScoreBoard(Turtle):
    def __init__(self):
        super().__init__()
        self.score = 0
        self.color("black")
        self.penup()
        self.hideturtle()

    def display_score(self):
        self.clear()
        ##self.penup()
        self.goto(0,250)
        self.write(f"Score: {self.score}", align="center", font=("Arial", 24, "normal"))

    def game_over(self):
        self.goto(0, 0)
        self.write("GAME OVER", align="center", font=("Arial", 24, "normal"))

    def increase_score(self):
        self.score += 1
        self.clear()
        self.display_score()


    def display_state(self, state, x, y):
        self.penup()
        self.goto(x,y)
        self.write(state, align="center", font=("Arial", 12, "normal"))

def game_play():
    counter =  0
    while counter <= 5:
        answer_state = screen.textinput(title = "Guess the state", prompt = "Another state's name")
        if answer_state.lower() in  data_states.state.str.lower().values:
            tkinter.messagebox.showinfo("Result", f"Correct, good job! You have {5-counter} lives left")
            score.increase_score()
            xcor = data_states[data_states.state.str.lower() == answer_state]["x"].iloc[0]
            ycor = data_states[data_states.state.str.lower() == answer_state]["y"].iloc[0]
            score_new = ScoreBoard()
            score_new.display_state(answer_state, xcor, ycor)
        else:
            tkinter.messagebox.showinfo("Result", f"Sorry, try again! You have {5-counter} lives left")
            counter += 1
            
        


score = ScoreBoard()
score.display_score()
game_play()

turtle.mainloop()
```
