<h3>Problem Description: Build a Banking System </h3>
Base Class: BankAccount

Attributes:
account_holder (name of the account holder)
balance (initial balance, default to 0)
Methods:
deposit(amount) - Add the amount to the balance.
withdraw(amount) - Subtract the amount from the balance if sufficient funds are available. Otherwise, display an error message.
check_balance() - Print the current balance.
Derived Classes:

SavingsAccount
Additional Attribute: minimum_balance (set to 1000).
Override withdraw() to ensure the balance does not go below the minimum balance. If it does, deny the withdrawal and display a message.
CurrentAccount
Additional Attribute: overdraft_limit (set to 5000).
Override withdraw() to allow the balance to go negative up to the overdraft limit. If the overdraft limit is exceeded, deny the withdrawal.
Polymorphism:

Write a function perform_transaction(account, transaction_type, amount) that:
Accepts an account object (SavingsAccount or CurrentAccount), a transaction type ("deposit" or "withdraw"), and the amount.
Performs the transaction based on the account type.


```python
class BankAccount():
    def __init__(self, name):
        self.account_holder = name
        self.balance = 0


    def deposit(self, deposit_amount):
        self.balance += deposit_amount
        print(f"Your new balance is {self.balance}")

    def withdraw(self, withdraw_amount):
        if self.balance >= withdraw_amount:
            print(f"Here is your {withdraw_amount} Rs")
            self.balance -= withdraw_amount
        else:
            print(f"You dont have enough money, you only have {self.balance} left in your account")


    def check_balance(self):
        print(f"Your current balance is {self.balance}")


class SavingsAccount(BankAccount):
    def __init__(self, name):
        super().__init__(name)
        self.minimum_balance = 1000


    def withdraw(self, withdraw_amount):
        if self.balance >= self.minimum_balance + withdraw_amount:
            print(f"Here is your {withdraw_amount} Rs")
            self.balance -= withdraw_amount
        else:
            print(f"Sorry you need minimum {self.minimum_balance} in your account ")


class CurrentAccount(BankAccount):
    def __init__(self, name):
        super().__init__(name)
        self.overdraft_limit = 5000


    def withdraw(self, withdraw_amount):
        if withdraw_amount >= self.minimum_balance + self.overdraft_limit:
            print(f"Here is your {withdraw_amount} Rs")
            self.balance -= withdraw_amount
        else:
            print(f"Sorry you need minimum {self.minimum_balance} in your account ")



def perform_txn(account_type, txn_type, amount):
    if type(account_type) in [SavingsAccount, CurrentAccount]:
        if txn_type == "deposit":
            account_type.deposit(amount)
        elif txn_type == "withdraw":
            account_type.withdraw(amount)
        else:
            print(f"Pls enter correct txn type")
    else:
        print(f"Pls enter a relevant account type")


shubham = SavingsAccount("shubham")

perform_txn(shubham, "deposit", 1000)
perform_txn(shubham, "deposit", 1000)

perform_txn(shubham, "withdraw", 500)
            
            
            
```
