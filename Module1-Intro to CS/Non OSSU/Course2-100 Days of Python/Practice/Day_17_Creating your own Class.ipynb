<h3> Constructor for definiting attributes </h3>

```python
## Constructor to initialise a class and define the starting variables

class UserNew:
    def __init__(self, user_id, username):  ## used to initialise the value of a class and will be called everytime a new object from this class is created
        print("new user being created")
        self.id = user_id ##defining new attributes for the class (here the attribute name is id and it will take up value user_id which will be defined while creating a new object)
        self.username = username ## defining new attributes for the class
        self.followers = 0

user_1 = UserNew("001", "shubham") ## we are creating a new user and overwriting the original user_1

print(user_1.id)
```

```python
print(user_1)
```

```python
user_2 = UserNew() ##this throws an aerror because we have to deefine user_id and username attributes everytime a new object is to be created
```

```python
class UserMethod:

    def __init__(self, user_id, username):
        print("new user being created")
        self.id = user_id ##defining new attributes for the class (here the attribute name is id and it will take up value user_id which will be defined while creating a new object)
        self.username = username ## defining new attributes for the class
        self.followers = 0
        self.following = 0


    def follow(self, user): ## self is required to be passed an argument, and we can have more params
        user.followers += 1
        self.following += 1


user_1 = UserMethod("001", "shubham")
user_2 = UserMethod("002", "survi")

user_1.follow(user_2) ## Calling a method here for user_1 which is the self, and passing the user_2 as the second argument
```

```python
print(user_1.following)
print(user_1.followers)
print(user_2.following)
print(user_2.followers)
```
