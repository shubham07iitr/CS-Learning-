<h1> Building a Pong game </h1>

### Generate a screen 
- 600*600 screen
- My pong game, should be blac
### Create a line on the screen
- Static line with nothing majore
### Creating the scoreboard on screen
- Right side for the player and left side for computer and starts with 0-0
- Updates if the ball object goes beyond the screen for either of the player
### Creating the ball object
- Gets generated from the center of the screen
- Gets thrown to the user who lost a point at a certain angle
- Gets deflected from the top or bottom roof and floor
- Getes deflected if it meets the paddle
### Creating the paddle object
- 1 paddle move up and down at a certain speed - the computer object
- 1 paddle is controlled by keyboard - the player object and can only move up and down near the edge

```python
import time
import random
%gui tk
from turtle import Screen, Turtle
```

```python
## Creating Class Scorecard
## Getting the Score Board


class Scoreboard(Turtle):

    def __init__(self):
        super().__init__()
        self.player_score = 0
        self.computer_score = 0
        
        
    def display_score(self):
        self.color("white")
        self.penup()
        self.goto(-80,240)
        self.pendown()
        self.write(f"{self.player_score}", align="center", font=("Arial", 48, "normal"))
        self.penup()
        self.goto(+80,  240)
        self.pendown()
        self.write(f"{self.computer_score}", align="center", font=("Arial", 48, "normal"))
        self.hideturtle()


    def score_increase(self):
        self.player_score += 1
        self.computer_score += 1
```

```python
## Creating the paddle class

class Paddle(Turtle):

    def __init__(self, p_type):
        super().__init__()
        valid_types = ['computer', 'player']
        if p_type not in valid_types:
            raise ValueError(f"Invalid p_type. Allowed values are: {valid_types}")
        self.p_type = p_type
        self.segment_player = []
        self.segment_computer = []


    def paddle_initialise(self):
        if self.p_type == "computer":
            for i in range(-2,3, 1):
                paddle_bit = Turtle(shape = "square")
                paddle_bit.color("white")
                paddle_bit.left(90)
                paddle_bit.penup()
                paddle_bit.goto(-430, i*20)
                self.segment_player.append(paddle_bit)
        else:
            for i in range(-2,3, 1):
                paddle_bit = Turtle(shape = "square")
                paddle_bit.color("white")
                paddle_bit.left(90)
                paddle_bit.penup()
                paddle_bit.goto(+430, i*20)
                self.segment_computer.append(paddle_bit)


    def paddle_movement_up():
        if self.p_type == "player":
            if self.segment_player[-1] or self.segment_player[0] >= abs[440]:
                pass
            else:
                for i in self.segment_player:
                    i.forward(40)
        else:
            pass


 
    def paddle_movement_down():
        if self.p_type == "player":
            if self.segment_player[-1] or self.segment_player[0] >= abs[440]:
                pass
            else:
                for i in self.segment_player:
                    i.backward(40)
        else:
            pass

```
