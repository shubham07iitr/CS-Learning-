- Figure out how to read separate lines in a text file
- Figure out how to replace a specific string in a file which was read
- Run a loop through all the separate lines in the names file and for each name create a new file

```python
import os

path_input_letter_base = os.getcwd() + '/Mail Merge Project Start/Input/Letters/starting_letter.txt'
path_input_names = os.getcwd() + '/Mail Merge Project Start/Input/Names/invited_names.txt'
path_output_mails = os.getcwd() + '/Mail Merge Project Start/Output/ReadyToSend/'



with open(path_input_names) as file:
    list_names = (file.readlines())
with open(path_input_letter_base) as file:
    letter_text = file.read()

for i in list_names:
    name = i.strip()  # Remove newline characters
    content = letter_text.replace("[name]", name)
    
    # Save the letter with a proper filename
    with open(path_output_mails + f'invite_letter_{name}.txt', 'w') as file:
        file.write(content)

print("Invitation letters created successfully!")
```
