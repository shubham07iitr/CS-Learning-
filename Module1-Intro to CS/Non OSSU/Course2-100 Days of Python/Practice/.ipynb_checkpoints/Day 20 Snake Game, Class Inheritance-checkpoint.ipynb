<h1> Snake game part 1 </h1>

```python
## Setting a black screen
## Get turtle to moving forward

#Get turtle to turn left and right

#Get turtle to die if it hits the edges of the screen

#Get turtle to eat food and get bigger

# Get turtle to to die if it hits its tali
import time
import random
%gui tk
from turtle import Screen, Turtle
screen = Screen()
screen.setup(width = 600, height = 600)
screen.bgcolor("black")
screen.title("My snake game")
screen.tracer()


## Creating a writer when you lose the game

writer = Turtle()
writer.hideturtle()  # Hide the turtle cursor
writer.penup()       # Lift the pen so it doesn't draw lines

# Display a message
def display_message(color="white", font=("Arial", 24, "normal")):
    writer.goto(0,0)  # Move to the specified position
    writer.color(color)    # Set the text color
    writer.write("Game Over", align="center", font=font)


## Creating a basic snake object
starting_position = [(0,0), (-20,0),(-40,0)]
segments = []
for position in starting_position:
    timmy = Turtle(shape = "square")
    timmy.color("white")
    timmy.penup()
    timmy.goto(position)
    segments.append(timmy)
    print(timmy.shapesize())



##to get precision for adding snake tail
def is_close(pos1, pos2):
    return abs(pos1 - pos2) < 1e-6



## Defining the food object
def food_pop():
    screen.tracer(0)
    food = Turtle(shape = "circle")
    food.color("white")
    food.shapesize(0.5, 0.5) ## Defining the size to be half of the segments
    food.penup()
    food.goto(random.randint(-13,13)*20, random.randint(-13, 13)*20)
    screen.tracer(1)
    print(food.shapesize())
    return food


def food_disappear(food):
    food.hideturtle()
    food.clear()

##Function to add another block at the end 
def snake_tail_new():
    screen.tracer(0)
    new_x = segments[-1].xcor() - 20 
    new_y = segments[-1].ycor() - 20
    timmy = Turtle(shape = "square")
    timmy.color("white")
    timmy.penup()
    timmy.goto(new_x, new_y)
    screen.tracer(1)
    segments.append(timmy)


## Defining the movement function where the last block follows the previous block and the head moves forward
def move():
    """Move the snake forward."""
    for i in range(len(segments) - 1, 0, -1):
        new_x = segments[i - 1].xcor()
        new_y = segments[i - 1].ycor()
        segments[i].goto(new_x, new_y)
    segments[0].forward(20)


## Defining the rotation of each of the blocks

def turn_right():
    if int(segments[0].heading()) == 90:
        for i in segments:
            screen.tracer(0)
            i.right(90) 
            screen.tracer(1)
    elif int(segments[0].heading()) == 270:
        for i in segments:
            screen.tracer(0)
            i.left(90)
            screen.tracer(1)


def turn_left():
    if int(segments[0].heading()) == 90:
        for i in segments:
            screen.tracer(0)
            i.left(90) 
            screen.tracer(1)
    elif int(segments[0].heading()) == 270:
        for i in segments:
            screen.tracer(0)
            i.right(90)
            screen.tracer(1)
            

def turn_up():
    if int(segments[0].heading()) == 0:
        for i in segments:
            screen.tracer(0)
            i.left(90) 
            screen.tracer(1)
    elif int(segments[0].heading()) == 180:
        for i in segments:
            screen.tracer(0)
            i.right(90)
            screen.tracer(1)


def turn_down():
    if int(segments[0].heading()) == 0:
        for i in segments:
            screen.tracer(0)
            i.right(90) 
            screen.tracer(1)
    elif int(segments[0].heading()) == 180:
        for i in segments:
            screen.tracer(0)
            i.left(90)
            screen.tracer(1)


## Interacting with the keyboard
screen.listen()
screen.onkey(turn_right, "Right")
screen.onkey(turn_left, "Left")
screen.onkey(turn_up, "Up")
screen.onkey(turn_down, "Down")

##Main logic
food_new = food_pop()
while True:
    screen.update()
    time.sleep(0.05)
    if abs(segments[0].xcor()) >= 280 or abs(segments[0].ycor()) >= 280:  ## Logic for ending the loop when head catches a wall
        display_message()
        break
    elif is_close(segments[0].xcor(), food_new.xcor()) and is_close(segments[0].ycor(), food_new.ycor()):
        snake_tail_new()
        food_disappear(food_new)
        food_new = food_pop()
    elif any(int(segments[0].xcor()) == int(segments[i].xcor()) and int(segments[0].ycor()) == int(segments[i].ycor()) for i in range(1, len(segments))):
        display_message()  # Display "Game Over"
        break
            
    else:
        move()
```

<h3> Class inheritance </h3> 

```python
## A new class inheriting some of its methods or attributes from a parent or super class

class Animal:
    def __init__(self):
        self.num_eyes = 2


    def breathe(self):
        print("Inhale, Exhale")


class Fish(Animal):

    def __init__(self):
        super().__init__() ## This will inherit all attributes and methods from the super Animal class

    def breathe(self):
        super().breathe() ## Calling the breathe method form the class Animal and modifying it to add one more print
        print("Doing this underwater")
    
    def swim(self):
        print("Moving in water")

fish = Fish()
fish.swim()
print(fish.num_eyes)
fish.breathe()
```
