```python
def odd_or_even(number):
    if number % 2 == 0:
        return "This is an even number."
    else:
        return "This is an odd number."

odd_or_even(20)
```

```python
# Target is the number up to which we count
def fizz_buzz(target):
    for number in range(1, target + 1):
        if number % 3 == 0 and number % 5 == 0:
            print("FizzBuzz")
        elif number % 3 == 0:
            print("Fizz")
        elif number % 5 == 0:
            print("Buzz")
        else:
            print(number)

fizz_buzz(5)
```

<h1> Day 15 Local Development Envirmonment Setup and Coffee Machine</h1>

```python
menu = {
    "espresso": {
        "ingredients": {
            "water": 50,
            "coffee": 18,
            'milk':0,
        },
        "cost": 1.5,
    },
    "latte": {
        "ingredients": {
            "water": 200,
            "milk": 150,
            "coffee": 24,
        },
        "cost": 2.5,
    },
    "cappuccino": {
        "ingredients": {
            "water": 250,
            "milk": 100,
            "coffee": 24,
        },
        "cost": 3.0,
    }
}

resources = {
    "water": 300,
    "milk": 200,
    "coffee": 100,
}

```

```python
#prompt the user 

money = 0
def user_prompt():
    user_input = input("What would you like , es, ca, la")
    return user_input

## generate report

def report():
    print(f"water: {resources["water"]}")
    print(f"milk: {resources["milk"]}")
    print(f"coffee: {resources["coffee"]}")
    print(f"money: {money}$")


##Asking user to insert coins
def txn_money():
    print(f"Pls insert coins")
    quarters = int(input("How many quarters:"))
    dimes = int(input("How many dimes:"))
    nickels = int(input("How many nickels:"))
    pennies = int(input("How many pennies:"))
    return quarters, dimes, nickels, pennies


def money_validation(user_input, quarters, dimes, nickels, pennies):
    if user_input == "es":
        if (quarters*0.25 + dimes*0.1+nickels*0.05 + pennies*0.01) >= 1.5:
            return True
        else:
            print("Sorry there is not enough money, pls try agian")
            return False
    elif user_input == "la":
        if (quarters*0.25 + dimes*0.1+nickels*0.05 + pennies*0.01) >= 2.5:
            return True
        else:
            print("Sorry there is not enough money, pls try agian")
            return False
    elif user_input == "ca":
        if (quarters*0.25 + dimes*0.1+nickels*0.05 + pennies*0.01) >= 3.0:
            return True
        else:
            print("Sorry there is not enough money, pls try agian")
            return False
        

def resources_validation(user_input):
    if user_input == "es":
        if (resources["water"] >= 50 and resources["coffee"] >= 18):
            return True
        else:
            for i in resources:
                if resources[i] < menu["espresso"]["ingredients"][i]:
                    print(f"Sorry there is not enough {i}: {resources[i]}")
                else:
                    pass
            return False

    elif user_input == "la":
        if (resources["water"] >= 200 and resources["coffee"] >= 24 and resources["milk"] >= 150):
            return True
        else:
            for i in resources:
                if resources[i] < menu["latte"]["ingredients"][i]:
                    print(f"Sorry there is not enough {i} : {resources[i]}")
                else:
                    pass
            return False

    elif user_input == "ca":
        if (resources["water"] >= 250 and resources["coffee"] >= 24 and resources["milk"] >= 100):
            return True
        else:
            for i in resources:
                if resources[i] < menu["cappuccino"]["ingredients"][i]:
                    print(f"Sorry there is not enough {i}: {resources[i]}")
                else:
                    pass
            return False

def post_positive_check_message(user_input, quarters, dimes, nickels, pennies):
    if user_input == "es":
        print(f"Here is your espresso, enjoy")
        print(f" here is {(quarters*0.25 + dimes*0.1+nickels*0.05 + pennies*0.01) - 1.5} $ in change")
    elif user_input == "la":
        print(f"Here is your latte, enjoy")
        print(f" here is {(quarters*0.25 + dimes*0.1+nickels*0.05 + pennies*0.01) - 2.5} $ in change")
    elif user_input == "ca":
        print(f"Here is your cappucino, enjoy")
        print(f" here is {(quarters*0.25 + dimes*0.1+nickels*0.05 + pennies*0.01) - 3.0} $ in change")
        


def resources_adjust(user_input, money_check, resources_check,money):
        if user_input == "es" and resources_check is True and money_check is True:
            resources["water"] -= 50
            resources["coffee"] -= 18
            resources["milk"] -= 0
            money += 1.5
        elif user_input == "la" and resources_check is True and money_check is True:
            resources["water"] -= 200
            resources["coffee"] -= 24
            resources["milk"] -= 150
            money += 2.5
        elif user_input == "ca" and resources_check is True and money_check is True:
            resources["water"] -= 250
            resources["coffee"] -= 24
            resources["milk"] -= 100
            money += 3.0
        else:
            return
            
        
    
    
```

```python
## defining game_logic

def game_logic():
    coffee_counter = False
    money_check = False
    resources_check = False
    money = 0
    print(resources)
    while coffee_counter is False:
        user_input = user_prompt()
        if user_input == "off":
            coffee_counter = True
            break
        elif user_input == "report":
            report()
        elif user_input in ["es", "ca", "la"]:
            quarters, dimes, nickels, pennies = txn_money()
            money_check = money_validation(user_input, quarters, dimes, nickels, pennies)
            resources_check = resources_validation(user_input)
            if money_check is True and resources_check is True:
                post_positive_check_message(user_input, quarters, dimes, nickels, pennies)
                print(resources)
                resources_adjust(user_input, money_check, resources_check, money)
                print(resources)
                
            else:
                pass
        else:
            print("Not an acceptable value, pls try again")


game_logic()
            
            
        
        
        
        

```
