<h3> HTTP Post Requests</h3>

- In Get requests we ask the external system to give data to us
- In Post requests we send new data to external system and not really too much interested in what they send back to us
- In put requests we update data in external system
- In delete we delete the data in external system
- Pixela documentation here
    - Creating a user - https://docs.pixe.la/entry/post-user
    - Creating a graph - https://docs.pixe.la/entry/post-graph
    - Posting a pixel - https://docs.pixe.la/entry/post-pixel
    - Update a pixel - https://docs.pixe.la/entry/put-pixel
    - Delet a pixel - https://docs.pixe.la/entry/delete-pixel

```python
import requests
pixela_endpoint = "https://pixe.la/v1/users"

USER_NAME = "shubham07iitr"
TOKEN = "doiwef0923uldkgdiwe"

user_params = {
    "token": TOKEN, ## some random otken which I created
    "username": USER_NAME,
    "agreeTermsOfService": "yes",
    "notMinor": "yes"}


response = requests.post(url = pixela_endpoint, json = user_params)
print(response.json())
    
```

<h3> Advanced authentication using HTTP Header </h3>

```python
graph_endpoint = f"{pixela_endpoint}/{USER_NAME}/graphs"
ID = "graph1"
graph_config = {
    "id": ID,
    "name": "Cycling Graph",
    "unit": "Km",
    "type": "float",
    "color": "ajisai"
}


## Now some web pages use headers to pass through API keys as this is more secure way 
headers = {
    "X-USER-TOKEN": TOKEN
}

response = requests.post(url = graph_endpoint, json = graph_config, headers = headers)
response.text

##our pizela graph can be seen here 
##https://pixe.la/v1/users/shubham07iitr/graphs/graph1
```

<h3> Posting a pixel </h3>

```python
pixel_endpoint = f"{pixela_endpoint}/{USER_NAME}/graphs/{ID}"

DATE = "20250314"
QUANTITY = "15.6"

headers = {
    "X-USER-TOKEN": TOKEN
}

pixel_config = {
    "date": DATE,
    "quantity": QUANTITY,
}
                

response = requests.post(url = pixel_endpoint, json = pixel_config, headers = headers)
response.text
```

<h3> Update and Delete pixels </h3> 

- Update a pixel - https://docs.pixe.la/entry/put-pixel
- Delet a pixel - https://docs.pixe.la/entry/delete-pixel

```python
pixel_update_endpoint = f"{pixela_endpoint}/{USER_NAME}/graphs/{ID}/{DATE}"



QUANTITY_UPDATED = "12.76"
update_config = {
    "quantity": QUANTITY_UPDATED}

headers = {
    "X-USER-TOKEN": TOKEN
}

##Updating
response = requests.put(url = pixel_update_endpoint, json = update_config, headers = headers)
response.text



```

```python
DATE_DELETE  = "20250315"
pixel_delete_endpoint = f"{pixela_endpoint}/{USER_NAME}/graphs/{ID}/{DATE_DELETE}"

## Deleting
response = requests.delete(url = pixel_delete_endpoint, headers = headers)
response.text
```
