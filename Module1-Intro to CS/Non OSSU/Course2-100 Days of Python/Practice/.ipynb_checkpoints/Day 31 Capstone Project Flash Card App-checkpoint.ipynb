```python
## Reading the file for french words:
import pandas as pd
import random
import time

def generate_word_wrong():
    df = pd.read_csv("Day31/data/french_words.csv")
    row = random.randint(0, df.shape[0] - 1)  # Pick one row
    french = df.loc[row, "French"]
    english = df.loc[row, "English"]
    return french, english


def generate_word_right():
    df = pd.read_csv("Day31/data/french_words.csv")
    word_drop = canvas.itemcget(word_text, "text")
    df.drop(df[df["French"] == word_drop].index, inplace=True)  # Drop row
    df.reset_index()  # Reset index
    row = random.randint(0, df.shape[0] - 1)  # Pick one row
    french = df.loc[row, "French"]
    english = df.loc[row, "English"]
    return french, english
    


def flip_canvas_right():  
    words = generate_word_right()
    canvas.itemconfig(word_text, text=words[0])  # Show French word
    canvas.itemconfig(language_text, text="French")
    canvas.tag_raise(french_id)  # Bring French card to front
    canvas.tag_raise(word_text)  # Bring text above image
    canvas.tag_raise(language_text)  # Bring text above image
    
    # After 3 seconds, flip to English word and card
    window.after(3000, lambda: (
        canvas.itemconfig(word_text, text=words[1]),
        canvas.itemconfig(language_text, text="English"),
        canvas.tag_raise(english_id),  # Bring English card to front
        canvas.tag_raise(word_text),  # Bring text above image
        canvas.tag_raise(language_text)  # Bring text above image

    
    ))
    french_word = words[0]
    df = pd.read_csv("Day31/data/french_words.csv")
    french_word = df[df["French"] == french_word]["French"].item()
    with open("flash_card", "a") as file:
        file.write(f"{french_word}: Know\n")

def flip_canvas_wrong():  
    words = generate_word_wrong()
    canvas.itemconfig(word_text, text=words[0])  # Show French word
    canvas.itemconfig(language_text, text="French")
    canvas.tag_raise(french_id)  # Bring French card to front
    canvas.tag_raise(word_text)  # Bring text above image
    canvas.tag_raise(language_text)  # Bring text above image
    canvas.itemcget(word_text, "text")
    # After 3 seconds, flip to English word and card
    window.after(3000, lambda: (
        canvas.itemconfig(word_text, text=words[1]),
        canvas.itemconfig(language_text, text="English"),
        canvas.tag_raise(english_id),  # Bring English card to front
        canvas.tag_raise(word_text),  # Bring text above image
        canvas.tag_raise(language_text)  # Bring text above image

    
    ))
    french_word = words[0]
    df = pd.read_csv("Day31/data/french_words.csv")
    french_word = df[df["French"] == french_word]["French"].item()
    with open("flash_card", "a") as file:
        file.write(f"{french_word}: Don't Know\n")

    
```

```python
# ---------------------------- CONSTANTS ------------------------------- #
GREEN_WINDOW = "#9bdeac"
GREEN_CANVAS = "#808000"

# ---------------------------- AESTHETICS & DESIGN ------------------------------- #

## Creating a new windown and title with bg = light green, title  = Flashy, size = 900,900

from tkinter import *
from tkinter import messagebox

window = Tk()
window.title("Flashy")
window.minsize(width = 800, height = 600)
window.config(padx = 140, pady = 100, bg = GREEN_WINDOW)


## Loading the french and  image on the screen
french_screen = PhotoImage(file = "Day31/images/card_front.png")
english_screen = PhotoImage(file = "Day31/images/card_back.png")
canvas = Canvas(width = french_screen.width(), height = french_screen.height(), bg = GREEN_WINDOW, highlightthickness=0) ##defining the canvas widget

french_id = canvas.create_image(french_screen.width() // 2, french_screen.height() // 2, image = french_screen) ## inserting the image in the canvas
english_id = canvas.create_image(english_screen.width() // 2, english_screen.height() // 2, image = english_screen) ## inserting the image in the canvas
canvas.tag_raise(french_id) 

language_text = canvas.create_text(french_screen.width() // 2,french_screen.height()//2 - 100, text = "French", fill = "black", font = ("Arial", 24, "italic")) ##inserting the french texh
word_text = canvas.create_text(french_screen.width() // 2,french_screen.height()//2, text = "Test", fill = "black", font = ("Arial", 48, "bold")) ##inserting the french texh
canvas.grid(column = 1, row = 2)


## Loading the wrong and right buttons on screen

wrong = PhotoImage(file = "Day31/images/wrong.png")
right = PhotoImage(file = "Day31/images/right.png")

wrong_button = Button(image=wrong, borderwidth=0, highlightthickness=0, bg=GREEN_WINDOW, activebackground=GREEN_WINDOW, command = flip_canvas_wrong)
right_button = Button(image=right, borderwidth=0, highlightthickness=0, bg=GREEN_WINDOW, activebackground=GREEN_WINDOW, command = flip_canvas_right)
wrong_button.grid(column = 1, row = 3, sticky = "w", padx=(120, 0))
right_button.grid(column = 1, row = 3, sticky = "e", padx=(0, 120))

flip_canvas_wrong()


window.mainloop()
```
