<h3> Building a quiz GUI project</h3>

```python
##building the check answer function
import random
import requests
from tkinter import messagebox
import html

class QuizBrain:
    def __init__(self):
        self.question_bank = []
        self.questions_answered =  0
        self.questions_correct = 0
        self.questions_incorrect = 0


    def generate_question_bank(self):
        response = requests.get(url = "https://opentdb.com/api.php?amount=10&type=boolean")
        self.question_bank = []
        for i in response.json()["results"]:
            self.question_bank.append((i["question"], i["correct_answer"]))
        return self.question_bank


    def generate_random_question(self):
        if self.question_bank:
            random_index = random.randint(0, len(self.question_bank) - 1)
            question = html.unescape(self.question_bank[random_index][0]) ## to avoid html entities screwing up our text
            canvas.itemconfig(question_text , text = question)
        else:
            pass
        

    
    def reset_bg(self):
        canvas.config(bg = "white")
    
    def check_answer(self, answer):
        if self.question_bank:
            question = canvas.itemcget(question_text, "text")
            index = 0
            found = False
            for i in self.question_bank:
                if i[0] == question and i[1] == answer:
                    self.questions_correct += 1
                    self.questions_answered += 1
                    score_label.config(text=f"Score: {self.questions_correct}/{self.questions_answered}")
                    canvas.config(bg = "green")
                    window.after(500, self.reset_bg)
                    found = True
                    break
                elif i[0] == question and i[1] != answer:
                    self.questions_incorrect += 1
                    self.questions_answered += 1
                    score_label.config(text=f"Score: {self.questions_correct}/{self.questions_answered}")
                    canvas.config(bg = "red")
                    window.after(500, self.reset_bg)
                    found = True
                    break
                else:
                    index += 1
                    pass
            if found:
                del self.question_bank[index]
                self.generate_random_question()
        else:
            right_button.config(state = disabled)
            wrong_button.config(state = disabled)
            messagebox.showinfo(title = "Final Result", message  = f"Quiz over, you answered {self.questions_correct} correctly of 10")

    
```

```python
##Building a GUI interface

from tkinter import *
from tkinter import messagebox

window = Tk()
window.title("Quizler")
window.config(width = 500, height = 800, bg = "#375362", padx = 100, pady = 20)


##Making the Score Label

score_label = Label(text = "Score: 0/0", bg = "#375362", fg = "white",  font = ("Arial", 12, "bold"))
score_label.grid(column = 1, row = 1, pady = 20)

## Making the grid
canvas = Canvas(width = 300, height = 300, bg = "white")
question_text = canvas.create_text(150, 150, text = "Questions will appear here, click on Start button",width=280,  font = ("Arial", 24, "bold"), fill ="black")
canvas.grid(row = 2, column = 1)

quiz = QuizBrain()

quiz.generate_question_bank()
quiz.generate_random_question()

## Making the buttons 
right = PhotoImage(file = "Day34Quizzler/images/true.png")
wrong = PhotoImage(file = "Day34Quizzler/images/false.png")
right_button = Button(image =  right, command = lambda: quiz.check_answer("True"))
wrong_button = Button(image =  wrong, command = lambda: quiz.check_answer("False"))
wrong_button.grid(column = 1, row = 4, sticky = "w", pady = 20)
right_button.grid(column = 1, row = 4, sticky = "e", pady = 20)





window.mainloop()
```
