Create a screen

Create a turtle class and object 
- Which can only move up through the keyboard

Display Level : 1 or two on top left

Creae another class cars which generates cars randomly 
- Generate a random no of cars between x and y coordinate to begin with
- Generate new cars after every x seconds from right side
- Cars should be xx size and to be picked from random colors

Detect collission logic b/w turtle and car and push game over

If turtle reaches at the end on y axis then:
- Turtle to go back to starting point
- Speed of cars to increase
- Fresh new cars to start
- Scorboard should display a new score
  

```python
import time
import random
%gui tk
from turtle import Screen, Turtle

##game_running = True
screen = Screen()
screen.setup(width = 600, height = 600)
screen.bgcolor("white")
screen.title("Turtle Crossing Game")

class TurtleObject(Turtle):
    def __init__(self):
        super().__init__()
        self.shape("turtle")
        self.color("black")
        self.left(90)
        self.penup()
        self.goto(0, -250)


    def move_up(self):
        if self.ycor() <=250:
            self.goto(self.xcor(), self.ycor()+25)


    def move_down(self):
        if self.ycor() >=-250:
            self.goto(self.xcor(), self.ycor()-25)


class Scoreboard(Turtle):

    def __init__(self):
        super().__init__()
        self.level = 1
        self.color("black")
        self.penup()
        self.hideturtle()
        
    def display_score(self):
        self.clear()
        ##self.penup()
        self.goto(-250,250)
        self.write(f"Level: {self.level}", align="center", font=("Arial", 24, "normal"))


    def game_over(self):
        self.goto(0, 0)
        self.write("GAME OVER", align="center", font=("Arial", 24, "normal"))

    def increase_score(self):
        self.level += 1
        self.clear()
        self.display_score()



class Car(Turtle):
    def __init__(self, y_position, color):
        super().__init__()
        self.shape("square")
        self.color(color)
        self.penup()
        self.goto(250, y_position)
        self.move_speed = 500
        self.shapesize(stretch_wid=1, stretch_len=2)

    def move_left(self):
        self.goto(self.xcor()-25, self.ycor())
        ##screen.ontimer(self.move_left, self.move_speed)


    @classmethod
    def create_random_cars(cls):
        """Creates a random number (1-4) of cars at different Y positions"""
        random_no_cars = random.randint(1, 4)
        color_list = ["red", "blue", "black" , "yellow", "green"]
        cars_list = []
        for _ in range(random_no_cars):
            random_ycor = random.randint(-20, 20)  # Spread cars vertically
            random_color = random.choice(color_list)
            car = cls(random_ycor*10, random_color)  # Pass Y position to constructor
            cars_list.append(car)
        return cars_list
        


        


game_is_on = True
screen.tracer(0)
turtle = TurtleObject()
score =  Scoreboard()
score.display_score()
screen.listen()
screen.onkey(turtle.move_up, "Up")
screen.onkey(turtle.move_down, "Down")
##cars_random = Car.create_random_cars()
cars_random = []
speed = 0.5
while game_is_on:
    screen.update()
    time.sleep(speed)
    for car in cars_random:
        car.move_left()
    if any(turtle.distance(car) <= 25 for car in cars_random):  # Increased distance to ensure collision detection
        score.game_over()
        game_is_on = False   # Stop execution
    if random.randint(1, 2) == 1:  # Random chance to generate new cars
        cars_random.extend(Car.create_random_cars())
    if turtle.ycor() >= 250:
        score.increase_score()
        turtle.penup()
        turtle.goto(0, -250)
        speed = speed/2
        
    
    
screen.exitonclick()
```
