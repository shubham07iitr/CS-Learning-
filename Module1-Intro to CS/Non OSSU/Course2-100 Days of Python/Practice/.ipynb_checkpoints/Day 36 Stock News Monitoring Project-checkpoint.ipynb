<h2>Setting up the stock prices </h2>

```python
import requests
import datetime
from datetime import datetime, timedelta


def get_tesla_dod_change():
    URL = "https://www.alphavantage.co/query?"
    API_KEY_ALPHA = "4U9TUU8XPR7SEM8E"
    params_stock = {
        "function": "TIME_SERIES_DAILY",
        "symbol": "TSLA",
        "apikey": API_KEY_ALPHA,
        "outputsize": "compact"
    }
    yesterday = (datetime.now() - timedelta(days=1))
    day_before = (datetime.now() - timedelta(days=2))
    response = requests.get(url = URL, params = params_stock)
    
    data = response.json()["Time Series (Daily)"]
    dod_change_value = float(data[str(yesterday.date())]['4. close']) - float(data[str(day_before.date())]['4. close'])
    dod_change_percent = dod_change_value/(float(data[str(day_before.date())]['4. close']))
    # if abs(dod_change_percent) > 0.02:
    #     print(f"Tesla changed by {round(dod_change_percent*100, 1)}% by today")
    # else:
    #     print("No material change in Tesla today")
                                             
    return round(dod_change_percent*100, 1)
                                                
get_tesla_dod_change()
```

```python
def get_tesla_news():
    URL = "https://newsapi.org/v2/everything"
    API_KEY_NEWS = "ffde7c2135624bd1916e49e94c418242"
    COMPANY_NAME = "Tesla"
    SEARCHIN = "title"
    FROM_DATE = (datetime.now() - timedelta(days=2)).date()
    TO_DATE = (datetime.now() - timedelta(days=2)).date()
    LANGUAGE = "en"
    SORT_BY = "relevancy"
    PAGE_SIZE = 3
    
    
    
    params_news = {
        "apiKey": API_KEY_NEWS,
        "q": COMPANY_NAME,
        "searchIn": SEARCHIN,
        "from": FROM_DATE,
        "to": TO_DATE,
        "language": LANGUAGE,
        "sortBy": SORT_BY,
        "pageSize":  PAGE_SIZE}
    
    response = requests.get(url = URL, params = params_news)
    data = response.json()["articles"]
    
    final_news = []
    for i in data:
        title = i["title"]
        description = i["description"]
        final_news.append((title, description))
    return final_news
    
    
get_tesla_news()  
```

```python
def actual_message():
    price_change = get_tesla_dod_change()
    news_alerts = get_tesla_news()  
    final_message = f"Tesla price changed by {price_change}% today, top news headlines are below:\n1. {news_alerts[0][0]}\n2. {news_alerts[1][0]}\n3. {news_alerts[2][0]}"
    return final_message

print(actual_message())
```
