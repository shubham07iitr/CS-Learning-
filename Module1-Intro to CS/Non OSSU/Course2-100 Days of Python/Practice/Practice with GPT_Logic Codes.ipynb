<h3> 1. The Mysterious Number (Pattern Recognition & Math) Problem: </h3>

You are given a number N, and you need to find the smallest integer X such that when you multiply X by the sum of its digits, you get N.

Example:

Input: N = 36
Output: X = 4
Explanation: 4 * (4) = 36 is false, but 6 * (6) = 36 is true.
Hint: Try iterating over potential values of X and check if X * sum_of_digits(X) == N.



```python
##How to calculate the sum of digits of a number

input_no = int(input("Pls enter the number for which you want the mysterious counterpart"))

counter = 1
while True:
    sum_counter = 0
    counter_str = str(counter)
    for i in counter_str:
        sum_counter = sum_counter + int(i)
    if counter >= input_no:
        print("No such no exists")
        break
    elif counter*sum_counter == input_no:
        print(f"the mysterious no is {counter}")
        break
    else:
        counter = counter+1
```

<h3> 1. Find Special Numbers (Divisibility Check) </h3>

**Problem:**

Given a list of numbers, find all numbers that are divisible by 3 but not by 5 using list comprehension.

**Example:**

Input: [10, 15, 21, 30, 33, 45, 50, 60, 99]

Output: [21, 33, 99]

```python
input_list = [10, 15, 21, 30, 33, 45, 50, 60, 99]
```

```python
output_list = [num for num in input_list if num%3 == 0 and num%5 != 0]
output_list
```
