```python
import turtle as t
import random

def random_color():
    r = random.randint(0, 255)
    g = random.randint(0, 255)
    b = random.randint(0, 255)
    return (r, g, b)

# Setting up the turtle and screen
t.colormode(255)  # Ensures the turtle module supports RGB values in 0-255
timmy = t.Turtle()
timmy.pensize(5)
timmy.speed("fastest")

angle = [90, 180, 270, 360]

# Drawing with random colors
for i in range(50):
    timmy.color(random_color())  # Use the random color generated
    timmy.right(random.choice(angle))  # Choose a random direction
    timmy.forward(10)  # Move forward

screen = Screen()
screen.exitonclick()
```

```python
import colorgram

colors = colorgram.extract('image.jpg', 30)

rgb_colors = []
for i in colors:
    r = i.rgb.r
    g = i.rgb.g
    b = i.rgb.b
    tuple_colors = (r,g,b)
    rgb_colors.append(tuple_colors)

rgb_colors[0]
```

```python
import turtle as t
import random

t.colormode(255)
timmy = t.Turtle()
timmy.pensize(2)
timmy.speed("fastest")

counter = 0
for x in range(0,6):
    for i in range(0,5):
        timmy.pendown()
        timmy.color("white")
        timmy.fillcolor(rgb_colors[counter])
        timmy.begin_fill()
        timmy.circle(5)
        timmy.end_fill()
        timmy.penup()
        timmy.forward(20)
        counter = counter+1
    timmy.right(180)
    timmy.forward(100)
    timmy.left(90)
    timmy.forward(20)
    timmy.left(90)
    


    
```
