<h3> Canvas widget </h3>

- Allows you to layer things one on top of another

**Dynamic Typing**
- Change a variables datatype by changing content in variable
- For e.g. i can write a = 10 , it will be stored as an int, but then if I again write a = "10" it will be stored as string

```python
from tkinter import *
import math

# ---------------------------- CONSTANTS ------------------------------- #
PINK = "#e2979c"
RED = "#e7305b"
GREEN = "#9bdeac"
YELLOW = "#f7f5dd"
FONT_NAME = "Courier"
WORK_MIN = 1
SHORT_BREAK_MIN = 5
LONG_BREAK_MIN = 20
reps = 0
timer = None


# ---------------------------- TIMER RESET ------------------------------- # 
def reset_timer():
    window.after_cancel(timer)
    canvas.itemconfig(timer_text, text = "00:00")
    timer_label.config(text = "Timer")
    check_marks.config(text = "")
    global reps
    reps = 0 ##ensuring that we start from 0 again
    


# ---------------------------- TIMER MECHANISM ------------------------------- # 

# ---------------------------- COUNTDOWN MECHANISM ------------------------------- # 

# ---------------------------- UI SETUP ------------------------------- #

window = Tk()
window.title("pomodoro")
window.config(padx = 100, pady = 50, bg = YELLOW)


##writing the start timer function which will trigger only once we start the timer:

def start_timer():
    global reps
    reps += 1
    work_min_sec = WORK_MIN*60
    short_break_sec = SHORT_BREAK_MIN*60
    long_break_sec  = LONG_BREAK_MIN*60


    #if its 1st, 3rd, 5th, 7th reps:
    ## if its 8th rep 
    if reps%8 == 0:
        timer_label.config(text = "LONG BREAK", font = ("Arial", 24, "bold"), fg="green", bg=YELLOW)
        count_down(long_break_sec)
        
    ## if its multiple of 2 reps
    elif reps%2 == 0:
        count_down(short_break_sec)
        timer_label.config(text = "SHORT BREAK",  font = ("Arial", 24, "bold"), fg="blue", bg=YELLOW)
    else:
        count_down(work_min_sec)
        timer_label.config(text = "WORK", font = ("Arial", 24, "bold"), fg="red", bg=YELLOW)


##writing the functino to udpate the tomato timer 
def count_down(count):
    global timer ##using this to call the after function and then will use after_cancel in another functino to cancel this
    count_min = math.floor(count/60)
    count_seconds = count%60
    if count_seconds < 10:
        count_seconds = f"0{count_seconds}"
    canvas.itemconfig(timer_text, text = f"{count_min}:{count_seconds}")
    if count_seconds == 0: ##using dynamic typing to refine the visualisation 
        count_seconds = "00"
    if count > 0:
        
        timer = window.after(1000, count_down, count - 1) ## function will repeat after every 1 second , and we also have an ending condition
    else:
        start_timer() ##calling the start_timer again once one rep is completed 
        mark = ""
        work_session = math.floor(reps/2)
        for i in range(work_session): ##adding extra check mark for every 2 sessions (1 for work and 1 for break)
            mark += "âœ…"
        check_marks.config(text = mark)


##Adding the timer label
timer_label = Label(text = "Timer", font = ("Arial", 24, "bold"), fg="green", bg=YELLOW)
timer_label.grid(column =  1, row = 0)


##Adding the tomato image
tomato = PhotoImage(file = "Day 29_pomodoro/tomato.png")
canvas = Canvas(width = 200, height = 224, bg=  YELLOW, highlightthickness = 0) ##defining the canvas widget
canvas.create_image(103,112, image = tomato) ## inserting the image in the canvas
timer_text = canvas.create_text(103,130, text = "00:00", fill = "white", font = (FONT_NAME, 35, "bold")) ##inserting the clock for the tomato
canvas.grid(column = 1, row = 2)


##Adding the start and reset widgets/buttons

start_button =  Button(text = "Start" , fg="black", bg=YELLOW , bd=0, highlightthickness=0 , relief = "flat", command = start_timer) ##calling the start_timer function when button is clicked
start_button.grid(column = 0 , row = 3)
reset_button =  Button(text = "Reset" , fg="black", bg=YELLOW,  bd=0, highlightthickness=0, relief = "flat", command = reset_timer)
reset_button.grid(column = 2 , row = 3)

##check mark label

check_marks = Label(text = "", fg = GREEN, bg = YELLOW)
check_marks.grid(column = 1 , row = 3)

window.mainloop()
```

<h3> After module </h3>
- Since tkinter is an event driven system, hence running a while/for loop while running the mainloop funciton will not work
- Hence we want to use recursion + after function to get the job done

```python
from tkinter import *
window = Tk()
window.title("after")
window.config(padx = 100, pady = 50)


##see how we are intelligently using after and recursion to repeat things
def count_down(count):
    print(count)
    if count > 0:
        window.after(1000, count_down, count - 1) ## function will repeat after every 1 second , and we also have an ending condition



count_down(5)
window.mainloop()
```
