```python
##Converting text to relevant data points from nutritionx

import requests

def generate_exercise_data():
    APP_ID = "29fb091b" 
    API_KEY = "46f7f653a74967539893b9d58fc9394e"
    
    TEXT = input("Tell me which exercise you did today")
    
    URL = "https://trackapi.nutritionix.com/v2/natural/exercise"
    
    config = {"query": TEXT 
        
    }
    headers = {
        "x-app-id": APP_ID,
        "x-app-key": API_KEY
    }
    
    response = requests.post(url = URL, json = config, headers = headers)
    data = response.json()
    return data
```

```python
## Getting the data to gsheet
from datetime import datetime
from requests.auth import HTTPBasicAuth

def post_workout_sheet():
    ##Generating data to be posted
    data = generate_exercise_data()

    ##Assuming there could be multiple row items:
    for i in range(0,len(data["exercises"])):
        ##Identifying relevant data points
        DATE = datetime.today().strftime("%d/%m/%Y")
        TIME = datetime.today().time().strftime("%H:%M:%S")
        EXERCISE = data["exercises"][i]["name"]
        DURATION = round(data["exercises"][i]["duration_min"], 0)
        CALORIES = round(data["exercises"][i]["nf_calories"], 0)


        ##Setting up the basic Auth for security reasons
        basic = HTTPBasicAuth("Authorization", "Basic c2h1YmhhbTA3aWl0cjpBcHJAMjAxNg==")        
        
        ##Defining post params
        URL = "https://api.sheety.co/e12f89f23521808a30a08256a250686a/myWorkoutsShubham/workouts"
        
        sheet_config = {
            "workout": {
                "date": DATE,
                "time": TIME,
                "exercise": EXERCISE,
                "duration": DURATION,
                "calories": CALORIES
            }
        }
    
        response = requests.post(url = URL, json = sheet_config, auth = basic)
        response.text
    

post_workout_sheet()
```
