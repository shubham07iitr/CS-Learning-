<h3> How internet actually works </h3>

- Internet is just a long piece of wire which connects multiple computers
- Some of the computers are servers as they have to serve you all files and data you need 24*7
- Any computer user would use to access a server is called a client
- Webserver is like a giant library which can serve you any website all the time
- How it works:
    - when you type in the name of a website lets say google.com, first a message is triggered to your Internet Service Provider
    - Then ISP sends a trigger to DNS server which is essentially a phone book capturing the IP address of the google server
    - DNS server returns the IP address to our computer which then sends a request to that IP address and website is loaded

<h1> CSS Intro </h1>

**CSS - Cascading Style Sheets**

- Style applied is cascading  start from general move on to specific
- Style sheet - type of language - how things should look like in a website
- Earlier all websites looked very dry as CSS was not there so no way to style websites
- Then in 97 HTML 3.2 intrdoced some attributes such as font, color etc to make more exciting webpages
- But then using font and colors within HTML would have cluttered the reading of HTML program
- Hence CSS was developed/introduced in 97 to be separately attaced to HTML for styling


<h3> How to add CSS </h3>

- 3 ways to add CSS to HTML
    - Inline
        - Goes into same line as the HTML element inside the HTML tag of each individual elelment
        - for e.g <html style="background:blue"> </html>
        - Useful only if we want to add styling for a single elements
        - But not recommended as we have to push this into each and every element 
    - Internal
        - Inside style tag in HTML head and have to be defined onlye once for each element
        - We add < style> html {background:red;} < / style>
        - Affects only that page 
    - External
        - Lives in a completely separate file like main.css
        - We join the HTML file with CSS file by using a link element inside the <head>
        - And using the attributes - rel ="stylesheet", href="./styles.css"
        - Used most commonly in web development

<h3> CSS Selectors </h3>

- Element Selector - Applies to all elements which have that particular tag
    - For e.g. if we do h1{color: red} it will apply to all h1 elements
- Class selector - applies to only the class which was specified
    - For e.g. .red-heading {color: red} will only allow red color in class named red-heading
    - It has to start by a dot followed by the classname like .red-heading
    - Now how to define class - by including it as an attribute in the respective elements such as below
    - < h1 class = "red-text"> Red </h1> Now this can be applied to any combination and not only for h1 elements
- ID Selector
    - Has to be defined by #IDname{color: red}
    - We also have to specify the IDname as attribute in the element <h2 = "IDname">Red</h2>
    - Difference is IDs are unique and classes can have more than one elements
- Attribute selector
    - Styling all elelments which use a particular attribute
    - For e.g p[draggable]{color: red} will color red all para elements which have used draggable as the attribute
    - We can even specify that for which value of attribute, the elements should be styled
    - p[draggable="false"]{color="red"}
- Universal selector
    - Applies to all elements like *{color: red}
      
- If I define a universal selecgtor *{color: red} and also an id selector #main{color: blue} and I define one of my h2 hading attribute ID = main, now for that h2 element - it will be blue because **CSS selectors follow rule of higher specificty, that is id will take precedence over class over universal**

<h3> Font Properties </h3>

- h1 {color: blue}
- **Font weight** h1 {font-weight: bold}
    - normal or bold
    - Lighter bolder : relative to parent
    - number: 100-900 (where 100 is lightest)
- **Font size** h1 {font-size: 20px}
    - 1px or 1 pixel is 1/96th of an inch or 0.26m
    - 1pt or 1 point is 1/72th of an inch 0.35mm
    - Relative size
        - 1em is basically 100% the size of parent and 2em is 200% and so on
        - or 1rem is 100% of root element or usually the HTML element
- **Font Family** h1 {font-family: sans-serif}
    - h1 {font-family: helvetica, sans-serif} ## hwere sans-serif is a backup to helvetica if it is not present in the laptop
    - Can copy from fonts.google.com and paste in style after </style> closing element
- **text-align** h1{text-align: center}
    - Can take values such as left, right, or even start and end which specifies from where the start would be like from center or right

<h3>CSS Box Model</h3>

- **border syling** h1{border: 10px solid black;} - takes 3 arguments
    - 10px - Thickness, style - dashed or solid, color of border - black or any other coloe
    - can also push bordert-top separately like below
        - h1{
            - border: 10px solid black;
                - border-top: 0px} border-top has to be indented
- **border-width** - takes 4 values1st is top, then right then bottom and then lef - clockwise
      - border: 30px solid black;
      - border-width: 10px 20px 30px 0px
    - If we pass only two values 0px and 20px then 1 and 3 which is top and bottom will take 0px and right and left will take 20px
- **padding** - h1{padding: 20px} will push my border out by 20px
- **margin** - This is outside the border
    - h1 {margin: 10px 20px 30px 40px} Again takes 4 values
- We can also box two or more elements together to have similar styling for them theough the **< div>** element
    - which kind of acts like a container for all element inside it
    - <  div>
        - < h1> This is heading 1 < /h1>
        - < p> This is para < /p>
    -  < /div>
    -  This way we can style the div elelemts together
