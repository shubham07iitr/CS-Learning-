<h3> Rain Alert Application </h3>

- The above data responds with 40 different entries 5 days, 3 hour intervals so 8 entries per day - so 40 entries
- but what I want to build is check everyday at 6AM the forecast for next 12 hours and whether it will rain or not
- So technically we need only the first 4 entries and not the remaining 36 entries
- param is "cnt" or count which can give us only the required ount

```python
import requests
API_KEY = "81975c0b2bc51a2db98426bc499ebd4b" ##from openweather map
lat = 26.9124
long = 75.7873

params = {"lat": lat,
         "lon": long,
         "appid": API_KEY,
         "cnt":4}

response =  requests.get(url = "https://api.openweathermap.org/data/2.5/forecast", params = params)
response.raise_for_status()
weather_data = response.json()

def check_rain_forecast():
    for i in weather_data["list"]:
        if i["weather"][0]["id"] < 700:
            return True
            break
        else:
            pass

if check_rain_forecast():
    print("Bring an umbrella")
else:
    print("No forecast for rain")
    
```

<h3> Environment Variables </h3>

- Environment variables are dynamic values that are part of the environment in which a process runs. In Python, they are typically used to store configuration settings outside of your code â€” like API keys, database URLs, or secret credentials.
- Why use environment variables?
    - Security: Store sensitive info like passwords, secret keys outside the source code.
    - Flexibility: Change behavior without changing code (e.g., development vs. production settings).
    - Portability: Code runs on different machines with different configurations.
- **Important thing to note is they are not stored inside the main.py file but rather inside the terminal**
- api_key = os.getenv('API_KEY') - this is how we will access an environment vairable
- **Using only export API_KEY=123abc will only save the variable for that session in the terminal**
- **To permanently save the variable , we need to sve it in Zshrc**

--------------------------------------------------------------------------------------------------------------------------------------------
- echo 'export API_KEY="123abc"' >> ~/.zshrc
