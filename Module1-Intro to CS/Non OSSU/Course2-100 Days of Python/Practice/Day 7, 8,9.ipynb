
<h1> day 7  hangman project </h1>

```python
word_list = ["baboob", "camel", "shubham", "aardvark"]
```

```python
## generating a random word from the list

import random
word_position_list = random.randint(0,(len(word_list)-1))
print(word_position_list)
word_to_chose = word_list[word_position_list]
print(word_to_chose)

# generating equivalent no. of spaces from the word

word_length = len(word_to_chose)
print(word_length)

word_list_blank = list()
for i in range(1, word_length+1):
    word_list_blank.append("_")

#unpacking the word in a list
word_chosen_unpack_list = list()

a = 0
for i in range(0, len(word_to_chose)):
    word_chosen_unpack_list.append(word_to_chose[a])
    a = a+1

print(word_chosen_unpack_list)


```

```python
## Logic
counter = False
total_lives = 7
while counter is False:
    #asking user to input a word and checking if that letter is available in the word we chose
    input_word = input("Chose a letter")
    b = 1
    letter_found = False
    for i in word_chosen_unpack_list:
        if input_word == i:
            print(f'You found a right letter at position {b}')
            word_list_blank[b-1] = i
            letter_found = True
            print(word_list_blank)
            b = b+1
            letters_pending = 0
            for i in word_list_blank:
                if i == "_":
                    letters_pending += 1
                else:
                    pass
            if letters_pending == 0:
                print("Congratulations you won")
                counter =  True
                break
        else:
            b = b+1

    if counter is True:
        break

    total_lives = total_lives -1 
    counter_local = 0
    for i in word_list_blank:
        if i == "_":
            counter_local += 1
        else:
            pass

    print(f'You have {counter_local} letters to be filled and {total_lives} chances left')

    if total_lives > 0 and letter_found is False:
        print("Sorry letter not found, pls try again")
    elif total_lives > 0 and letter_found is True:
        print("Great, you found a letter/s , pls try another one")
    elif letters_pending == 0:
        pass
    else:
        print("Sorry you lost, game over")
        counter = True
    

```

```python
def calculate_love_score(name1, name2):
    # Combine and convert both names to lowercase
    combined = name1.lower() + name2.lower()
    
    # Characters to check for the love score
    true_love = "truelove"
    
    # Count occurrences of each character in 'true_love' within 'combined'
    count_t = sum(combined.count(char) for char in true_love)
    
    print(f"Your total love score is {count_t}")

calculate_love_score("true", "love")
```

<h2> Caesar Cipher </h2>

```python
letters = ["a", "b", "c", "d", "e", "f", "g", "h", "i", "j",  "k", "l", "m", "n", "o", "p", "q", "r", "s", "t", "u", "v", "w", "x", "y", "z"]
letters[25]
```

```python
def encrypt(user_word,shift):
    encryption = ""
    for i in user_word:
        if i == " ":
            encryption = encryption + " "
        for j in letters:
            if i == j:
                index_counter = letters.index(j) + shift
                if index_counter <= 25:
                    encryption = encryption + letters[index_counter]
                else:
                    index_counter = index_counter - 26
                    encryption = encryption + letters[index_counter]
            else:
                pass
            
    print(f'The new encoded message is {encryption}')
```

```python
encrypt("shubham gupta", 10)
```

```python
user_word = input("Pls chose a word to be encoded")
shift = int(input("pls chose the shift"))
encrypt(user_word, shift)
```

```python
def decrypt(user_word,shift):
    encryption = ""
    for i in user_word:
        if i == " ":
            encryption = encryption + " "
        for j in letters:
            if i == j:
                index_counter = letters.index(j) - shift
                if index_counter >= 0:
                    encryption = encryption + letters[index_counter]
                else:
                    index_counter = index_counter + 26
                    encryption = encryption + letters[index_counter]
            else:
                pass
            
    print(f'The decoded message is {encryption}')
```

```python
user_word = input("Pls chose a word to be encoded")
shift = int(input("pls chose the shift"))
decrypt(user_word, shift)
```

```python
def caesar(user_word, shift, user_choice):
    if user_choice == "encode":
        encrypted_message = encrypt(user_word, shift)
        ##print(f'The new encoded message is: {encrypted_message}')
    elif user_choice == "decode":
        decrypted_message = decrypt(user_word, shift)
        ##print(f'The decoded message is: {decrypted_message}')
    else:
        print("Please choose a correct input (encode or decode).")
```

```python
user_word = input("Pls chose a word to be encoded")
shift = int(input("pls chose the shift"))
user_choice = input("Pls chose encode or decode")
caesar(user_word, shift, user_choice)
```

<h3> grading exercise </h3>

```python
student_scores = {
    'Harry': 88,
    'Ron': 78,
    'Hermione': 95,
    'Draco': 75,
    'Neville': 60
}

student_grades  = {}
for i in student_scores:
    if student_scores[i] >= 91:
        student_grades[i] = "Outstanding"
    elif student_scores[i] >= 81 and student_scores[i] < 90:
        student_grades[i] = "Exceeds Expectations"
    elif student_scores[i] >= 71 and student_scores[i] < 80:
        student_grades[i] = "Acceptable"
    else:
        student_grades[i] = "Fail"

print(student_grades)
```
