```python
def computer_guess():
    print("Think of a no between 1 and 100, and I will guess it")
    guess_list = [1,100]
    counter = 0
    while True:
        user_response = input(f"Is it {int((sum(guess_list))/2)} (Enter 'too high', 'too low', or 'correct'):")
        if user_response.lower() == "too high":
            guess_list[1] = int((sum(guess_list))/2)
            counter = counter + 1
        if user_response.lower() == "too low":
            guess_list[0] = int((sum(guess_list))/2)
            counter = counter + 1
        if user_response.lower() == "correct":
            print(f"I guessed it in {counter+1} attempts")
            break
    return counter

    
```

```python
#Defining a class of Library object
class Library():

    def __init__(self, books):
        self.books = books


    def initial_prompt(self):
        print("Welcome to the library")
        print("1. Add a book")
        print("2. View all books")
        print("3. Borrow a book")
        print("4. Return a book")
        print("5. Exit")


    def add_books(self):
        book_name = input("Enter the name of the book you want to add")
        self.books.append(book_name)
        print(f"Book {book_name} has been added to the library")


    def view_books(self):
        print("Books in the Library")
        for i in self.books:
            print(i)


    def borrow_books(self):
        book_name = input("Enter the name of the book you want to borrow")  
        if book_name in self.books:
            print(f"Here is your copy of {book_name}, you have to return it in 14 days")
            self.books.remove(book_name)
            
        else:
            print(f"Sorry {book_name} book is currently not available, pls try something else")


    def return_books(self):
        book_name = input("Enter the name of the book you want to return")
        self.books.append(book_name)
        print(f"Thank you for returning {book_name}")

 
```

```python
import random
```

```python
## Initialising the Dice class

class Dice():

    def __init__(self, sides):
        self.sides = sides


    def roll(self):
        print(f"rolling a D{self.sides}")
        dice_roll = random.randint(1, self.sides)
        print(f"You rolled a {dice_roll}")
        return dice_roll


    def get_sides(self):
        return self.sides


    def roll_again(self):
        while True:
            response = input("Do you want to roll again, Y or N")
            if response in ["Y", "N"]:
                return response
            else:
                print("Invalid input , pls try again 'Y' or 'N'")


    def dice_logic(self):
        dice_list = []
        a = self.roll()
        dice_list.append(a)
        while True:
            check = self.roll_again()
            if check == "Y":
                b = self.roll()
                dice_list.append(b)
            else:
                print(f"Results of your roll {dice_list}")
                print(f"Total sum {sum(dice_list)}")
                break
        
```

<h1> Project: Text-Based RPG Battle Simulator </h1>
Objective
Create a program that simulates a turn-based battle between a player and an enemy. The user will:

Create a character (player).
Battle an enemy (computer).
Choose actions during the battle (e.g., attack, heal).
Win if the enemy’s health reaches 0 or lose if the player’s health reaches 0.
This project involves concepts like randomness, user decision-making, and OOP while being different from what you’ve done so far.

Details

Attributes:

Attributes:
name: The enemy’s name.
health: The enemy’s health points (starts at 100).

```python
import random
```

```python
## Initialising Player Class
class Player():
    def __init__(self, name):
        self.name = name
        self.health = 100
        self.attack_power = [10,20]

    def attack(self):
        attack_value = random.randint(self.attack_power[0], self.attack_power[1])
        return attack_value

    def heal(self):
        self.health = self.health + random.randint(10,30)
        return self.health

    def is_alive():
        if self.health > 0:
            return True


    def player_turn(self, enemy):
        print("Your turn\n 1. Attack \n 2. Heal")
        while True:
            try:
                user_choice = int(input("Enter your choice"))
                if user_choice in [1,2]:
                    break
            except ValueError:
                print("Please chose a value between 1 and 2 only")
        if user_choice == 1:
            attack_number = self.attack()
            print(f"You attacked {enemy.name} for {attack_number} damage")
            enemy.health -= attack_number
            print(f"{enemy.name} has {enemy.health} left now")

        else:
            heal_number = self.heal()
            print(f"You healed yourself for {heal_number}")
            print(f"Your current health is {self.health}")
                
            
        
        

```

```python
## Initialising Enemy Class

class Enemy():
    def __init__(self, name):
        self.name = name
        self.health = 100
        self.attack_power = [10,20]


    def attack(self):
        attack_value = random.randint(self.attack_power[0], self.attack_power[1])
        return attack_value


    def is_alive(self):
        if self.health > 0:
            return True


    def enemy_turn(self, player):
        print(f"{self.name}'s turn now")
        enemy_attack = self.attack()
        print(f"{self.name} attacked you for {enemy_attack} damage")
        player.health -= enemy_attack
        print(f"You have {player.health} health left now")



        
        
```

```python
def initial_prompt():
    print("Welcome to the RPG battle simulator")
    player_name = input("Please enter your name")
    enemy_name = input("Please enter enemy's name")
    print(f"{player_name} prepare to battle the fearsome {enemy_name} now")
    print("Battle starts now")
    return player_name, enemy_name

def check_health(player, enemy):
    if player.health <= 0:
        print(f" Your health is now {player.health} Sorry you lose, Goblin wins")
        return False
    elif enemy.health <= 0:
        print(f"{enemy.name}'s health is {enemy.health}, which is less than equal to zero, congratulations you win")
        return False
    else:
        return True
```

```python
## Game Logic
while True:
    names = initial_prompt()
    player = Player(names[0])
    enemy =  Enemy(names[1])
    while check_health(player, enemy) is True:
        player.player_turn(enemy)
        enemy.enemy_turn(player)
    
    while True:
        try:
            play_again = input("Do you want to play again Y or N")
            if play_again in ["Y", "N"]:
                break
        except ValueError:
            print("Please enter a valid input from Y or N")
    if play_again == "Y":
        continue
    else:
        break
            
```

```python
## initialising the class train

class TrainTickets():

    def __init__(self):
        self.available_trains = [{"Name": "Rajdhani Express", "Number": "12345", "Total Seats": 20, "Available Seats": 20, "Booked Seats": 0},
                                 {"Name": "Shatabdi Express", "Number": "54321", "Total Seats": 20, "Available Seats": 20, "Booked Seats": 0}]

    def view_trains(self):
        for i in self.available_trains:
            print(f"{i["Name"]} - Train no: {i["Number"]} - Available seats: {i["Available Seats"]}")     


    def book_tickets(self):
        while True:
            try:
                user_choice = input("Which train would you like to book a ticket in")
                if user_choice in [self.available_trains[0]["Number"], self.available_trains[1]["Number"]]:
                    break
            except ValueError:
                print("Please chose a valid train numbr between 12345, and 54321")

        for i in self.available_trains:
            if i["Number"] == user_choice and i["Available Seats"] > 0:
                i["Available Seats"] = i["Available Seats"] - 1
                i["Booked Seats"] = i["Booked Seats"] +1
                print(f"Ticket booked successfully! Available seats for {i["Name"]}: {i["Available Seats"]}")
            else:
                pass


    def cancel_tickets(self):
        while True:
            try:
                user_choice_cancel = input("Which train would you like to cancel a ticket in")
                if user_choice_cancel in [self.available_trains[0]["Number"], self.available_trains[1]["Number"]]:
                    break
            except ValueError:
                print("Please chose a valid train numbr between 12345, and 54321")


        for i in self.available_trains:
            if i["Number"] == user_choice_cancel and i["Booked Seats"] > 0:
                i["Booked Seats"] = i["Booked Seats"] - 1
                i["Available Seats"] = i["Available Seats"] +1
                print(f"Ticket cancelled successfully! Available seats for {i["Name"]}: {i["Available Seats"]}")
            else:
                pass
            
        


            
                                   


            
```

```python
## final logic

train = TrainTickets()
while True:
    user_choice_initial = initial_prompt_train()
    if user_choice_initial ==  1:
        train.view_trains()
    elif user_choice_initial == 2:
        train.book_tickets()
    elif user_choice_initial == 3:
        train.cancel_tickets()
    else:
        break
    
    
```
