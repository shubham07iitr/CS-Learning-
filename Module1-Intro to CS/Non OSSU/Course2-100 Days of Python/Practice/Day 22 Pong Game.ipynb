```python
import random
from turtle import Screen, Turtle

## Creating Class Scoreboard
class Scoreboard(Turtle):
    def __init__(self):
        super().__init__()
        self.player_score = 0
        self.computer_score = 0
        self.color("white")
        self.penup()
        self.hideturtle()

    def display_score(self):
        self.clear()
        self.goto(-100, 200)
        self.write(f"Player: {self.player_score}", align="center", font=("Arial", 24, "normal"))
        self.goto(100, 200)
        self.write(f"Computer: {self.computer_score}", align="center", font=("Arial", 24, "normal"))

    def display_winner(self, winner):
        self.goto(0, 0)
        self.write(f"{winner} Wins!", align="center", font=("Arial", 36, "bold"))

## Paddle Class
class Paddle:
    def __init__(self, x_pos):
        self.segments = []
        self.create_paddle(x_pos)

    def create_paddle(self, x_pos):
        for i in range(-2, 3):  # Create paddle with 5 segments
            segment = Turtle("square")
            segment.color("white")
            segment.penup()
            segment.goto(x_pos, i * 20)
            self.segments.append(segment)

    def move_up(self):
        if self.segments[-1].ycor() < 240:  # Prevent paddle from going off-screen
            for seg in self.segments:
                seg.goto(seg.xcor(), seg.ycor() + 20)

    def move_down(self):
        if self.segments[0].ycor() > -240:  # Prevent paddle from going off-screen
            for seg in self.segments:
                seg.goto(seg.xcor(), seg.ycor() - 20)

    def follow_ball(self, ball):
        if ball.ycor() > self.segments[-1].ycor() and self.segments[-1].ycor() < 240:
            self.move_up()
        elif ball.ycor() < self.segments[0].ycor() and self.segments[0].ycor() > -240:
            self.move_down()

## Ball Class
class Ball(Turtle):
    def __init__(self):
        super().__init__()
        self.shape("circle")
        self.color("white")
        self.penup()
        self.reset_ball()

    def reset_ball(self):
        self.goto(0, 0)
        self.setheading(random.choice([45, 135, 225, 315]))  # Random angle

    def move(self):
        self.forward(10)

    def bounce_y(self):
        self.setheading(360 - self.heading())  # Reverse vertical direction

    def bounce_x(self):
        self.setheading(180 - self.heading())  # Reverse horizontal direction

## Center Line
def center_line():
    for i in range(-280, 300, 40):
        segment = Turtle("square")
        segment.color("white")
        segment.shapesize(stretch_wid=1, stretch_len=0.1)
        segment.penup()
        segment.goto(0, i)

## Game Logic
def play_game():
    global game_running  # Declare the variable as global at the top of the function
    if not game_running:
        return

    ball.move()

    # Ball bounces off top and bottom walls
    if ball.ycor() > 280 or ball.ycor() < -280:
        ball.bounce_y()

    # Ball bounces off paddles
    if ball.xcor() > 420 and any(ball.distance(seg) < 50 for seg in player_paddle.segments):
        ball.bounce_x()

    if ball.xcor() < -420 and any(ball.distance(seg) < 50 for seg in computer_paddle.segments):
        ball.bounce_x()

    # Ball goes out of bounds
    if ball.xcor() > 450:
        score.computer_score += 1
        score.display_score()
        ball.reset_ball()

    if ball.xcor() < -450:
        score.player_score += 1
        score.display_score()
        ball.reset_ball()

    # End the game when a player reaches 10 points
    if score.player_score == 10:
        score.display_winner("Player")
        game_running = False
        return
    elif score.computer_score == 10:
        score.display_winner("Computer")
        game_running = False
        return

    # Computer paddle follows the ball
    computer_paddle.follow_ball(ball)

    # Refresh the screen for smooth animation
    screen.update()

    # Continue the game
    screen.ontimer(play_game, 20)

## Screen Setup
screen = Screen()
screen.setup(width=900, height=600)
screen.bgcolor("black")
screen.title("Pong Game")
screen.tracer(0)

# Create Game Elements
center_line()
score = Scoreboard()
player_paddle = Paddle(430)
computer_paddle = Paddle(-430)
ball = Ball()

# Update screen for initial rendering
screen.update()

# Key Bindings
screen.listen()
screen.onkey(player_paddle.move_up, "Up")
screen.onkey(player_paddle.move_down, "Down")

# Start Game
game_running = True  # Declare and initialize the global variable here
score.display_score()
play_game()

# Mainloop
screen.mainloop()
```
