<h3> etch a sketch game </h3>

```python
%gui tk
from turtle import Turtle, Screen

tim = Turtle()
screen = Screen()

def move_forwards():
    tim.forward(100)

def move_backwards():
    tim.backward(100)

def turn_left():
    tim.left(10)

def turn_right():
    tim_right(10)

def clear():
    tim.clear()
    tim.penup()
    tim.home()
    tim.pendown()

screen.listen() ##it will start listening from user event
screen.onkey(move_forwards, "w") ## we are passing move_forwards into another function or method and in that case we dont need to put in the parentheses
screen.onkey(move_backwards, "s")
screen.onkey(turn_left, "a")
screen.onkey(turn_right, "d")
screen.onkey(clear, "c")
screen.exitonclick()
```

<h3> Turtle racing game </h3>

```python
import random
```

```python
import random
%gui tk
from turtle import Turtle, Screen

screen = Screen()
screen.setup(height= 400, width= 500)
user_bet = screen.textinput(title = "Make your bet", prompt = "Which turtle will win the race: enter a color")


colors = ["red", "blue", "green", "black", "brown"]

def initial_position(colors):
    counter = 0
    turtles_dict = {}
    for i in colors:
        timmy = Turtle()
        timmy.shape("turtle")
        timmy.color(i)
        turtles_dict[f"turtle_{i}"] = timmy
        timmy.penup()
        timmy.goto(timmy.xcor(), timmy.ycor()+counter)
        counter = counter+30
        timmy.backward(240)
    return turtles_dict
 


def move_forward(turtles_dict):
    counter = 0  
    winner = ""
    temp_tracker = turtles_dict.copy()
    for i in turtles_dict:
        temp_tracker[i] = 0
    
    while True:
        if counter < 490:
            temp_list = []
            for i in turtles_dict:
                step_no = random.randint(0,10)
                turtles_dict[i].forward(step_no)
                temp_list.append(step_no)
                temp_tracker[i] += step_no
            screen.update()
            counter += max(temp_list)
            print(counter)
        else:
            temp_list_winner = []
            for i in temp_tracker:
                temp_list_winner.append(temp_tracker[i])
            for i in temp_tracker:
                if temp_tracker[i] == max(temp_list_winner):
                    winner = i
                    break
                else:
                    pass
            print(winner)    
            if winner.split("_")[1].lower() == user_bet.lower():
                print(f"Great, you predicted the winner accurately as {user_bet}")
            else:
                print(f"Sorry you predicted {user_bet}, actual winner was {winner}")
            
            break
    
  
x = initial_position(colors)
print(x)

move_forward(x)
screen.exitonclick()
```
