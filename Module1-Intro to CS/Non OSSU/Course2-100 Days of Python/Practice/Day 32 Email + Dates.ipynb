- Email SMTP is a module which is pre-bundled in python through which you can send emials

<h3> Automated Bday Wisher </h3>

```python
import pandas as pd

def email_send(email, name, letter):
    my_email = "shubham07iitr@gmail.com"
    password = "cifhsoqrbycnefnt" ## this is an app password which we have to 

    with  smtplib.SMTP("smtp.gmail.com") as connection: ## Creating a new objet to conect to our email providers SMTP email server , text in quotes is the olocation of our email providers SMTP server
        connection.starttls()
        ##tls = transfer layer securtity is  a way of securing our connection - encrypting our message and securing our connection
        connection.login(user = my_email, password = password)
        connection.sendmail(
            from_addr= my_email, 
            to_addrs= email,
            msg = f"Subject:Happy B'Day\n\n{letter}"
        )


def check_bday():
    df = pd.read_csv("Day32Bday/birthdays_updated.csv")
    today_day = dt.datetime.now().day
    today_month = dt.datetime.now().month
    day_check = df["day"] == today_day
    month_check = df["month"] == today_month
    output = df[day_check & month_check]
    if output.shape[0] > 0:
        dict_bday = {}
        for index, row in output.iterrows():
            dict_bday[row["name"]] = row["email"]
    return dict_bday
            
    

def random_letter(name):
    number = random.randint(1,3)
    with open(f"Day32Bday/letter_templates/letter_{number}.txt", "r") as file:
        mail_list = file.readlines()
        if mail_list[0] == "Hey [NAME]":
            mail_list[0] = f"Hey {name}"
        else:
            mail_list[0] = f"Dear {name}"
    return mail_list

def final_mailing():
    day_dict = check_bday()
    if day_dict:
        for key, value in day_dict.items():
            name = key
            email = value
            letter = ''.join(random_letter(key))
            email_send(name = name, email = email, letter = letter)


final_mailing()
        
```
