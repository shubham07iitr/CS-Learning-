**4. Write a Python program that accepts a word from the user and reverses it.**

```python
def reverse():
    user_input = input("Please select a string to be reversed")
    temp_list = [i for i in user_input]
    final_str = ""
    for i in range(0, len(temp_list)):
        final_str = final_str + temp_list[-1-i]
    return f"Reversed string is {final_str}"

reverse()
    
        
```

```python
def perfect_square(num):
    for i in range(2, int((num/2)+1)):
        if math.log(num,i) - int(math.log(num, i)) == 0:
            return True
        else:
            pass
    return False


perfect_square(35)
    
```

```python
def sum_zero(list1):
    if len(list1) < 3:
        print("List is too small")
        return []

    zero_list = []
    list1.sort()  # Sorting helps avoid duplicates
    
    for i in range(len(list1) - 2):  # Loop till third last element
        for j in range(i + 1, len(list1) - 1):
            for k in range(j + 1, len(list1)):
                if list1[i] + list1[j] + list1[k] == 0:
                    triplet = [list1[i], list1[j], list1[k]]
                    if triplet not in zero_list:  # Avoid duplicate triplets
                        zero_list.append(triplet)
    
    if not zero_list:
        print("No such combination of numbers are there in the list")
        return []
    
    return zero_list

# Test Case
print(sum_zero([-1, 0, 1, 2, -1, -4]))  
# Expected Output: [[-1, -1, 2], [-1, 0, 1]]
```

**7. Write a Python program to push all zeros to the end of a list.**

Input : [0,2,3,4,6,7,10]

Output : [2, 3, 4, 6, 7, 10, 0]

```python
def push0(list7):
    counter = 0
    new_list = []
    for i in list7:
        if i == 0:
            counter += 1
        else:
            new_list.append(i)
    for i in range(0, counter):
        new_list.append(0)
    return new_list

push0([0,2,3,4,6,7,0,10])
            
    
```

**3. Write a Python class Employee with attributes like emp_id, emp_name, emp_salary, and emp_department and methods like calculate_emp_salary, emp_assign_department, and print_employee_details.**

Sample Employee Data:

"ADAMS", "E7876", 50000, "ACCOUNTING"

"JONES", "E7499", 45000, "RESEARCH"

"MARTIN", "E7900", 50000, "SALES"

"SMITH", "E7698", 55000, "OPERATIONS"

Use 'assign_department' method to change the department of an employee.

Use 'print_employee_details' method to print the details of an employee.

Use 'calculate_emp_salary' method takes two arguments: salary and hours_worked, which is the number of hours worked by the employee. If the number of hours worked is more than 50, the method computes overtime and adds it to the salary. Overtime is calculated as following formula:

overtime = hours_worked - 50

Overtime amount = (overtime * (salary / 50))**

```python
class Employee:

    def __init__(self, name, id, salary, department):
        self.name =  name.upper()
        self.id = id
        self.salary =  salary
        self.department =  department.upper()


    def assign_department(self, new_department):
        print(f"The original department was {self.department}")
        self.department = new_department.upper()
        print(f"the new department is {self.department}")


    def employee_details(self):
        print(f"The name of the employee is {self.name}")
        print(f"The ID of the employee is {self.id}")
        print(f"The salary of the employee is {self.salary}")
        print(f"The department of the employee is {self.department}")


    def calculate_emp_salary(self, hours_worked):
        print(f"the base salary of the employee is {self.salary}")
        if hours_worked > 50:
            overtime = ((hours_worked - 50)/50)*self.salary
            print(f"the overtime paid is {overtime}")
        else:
            print("There was no overtime pay")


shubham = Employee(name = "Shubham", id = "E7876", salary = 50000, department = "Accounting")
shubham.assign_department("Research")
shubham.employee_details()
shubham.calculate_emp_salary(60)
        
```
