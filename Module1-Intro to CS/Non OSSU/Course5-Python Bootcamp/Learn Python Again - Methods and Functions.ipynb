```python
work_hours = [('A', 100), ('B',150), ('C', 200)]
```

```python
def employee_check(work_hours):
    current_max = 0
    employee_of_month = ''

    for employee, hours in work_hours:
        if hours > current_max:
            current_max = hours
            employee_of_month = employee
            
        else:
            pass

    return (employee_of_month, current_max)
```

```python
employee_check(work_hours)
```

```python
name, hours = employee_check(work_hours)
```

```python
def myfunc(**kwargs): #kwawrgs returns a dictionary
    print(kwargs)
    if 'fruit' in kwargs:
        print(f'My choice of fruits is {kwargs['fruit']}')
    else:
        print('I did not find any fruit here')
```

<h2> Lamda expressions , maps and filters

```python
def square(num):
    return num**2
```

```python
## If I have to use the function on a list ,I can use the map function

list1 = [1,2,3,4,5,6]
for i in map(square, list1):
    print(i)
```

```python
def splicer(mystring):
    if len(mystring)%2 == 0:
        return 'Even'
    else:
        return mystring[0]
    
```

```python
names = ['Andy', 'Eve', 'Sally']
```

```python
list(map(splicer, names))
```

```python
list(filter(splicer, names))
```

```python
## we can use a filter function to filter out specific elements from a list

def even_check(nums):
    return nums%2 == 0

list1 = [1,2,3,4,2]

list(filter(even_check, list1))
```

<h4>Local <br> Enclosing function locals - for functions within a function <br>
Global <br>
Builtin <h2>

```python
#Global
name = "This is a global string"

def greet():
    #enclosing
    name = "Sammy"
    
    def hello():
        #Local
        print("Hello " +  name) ##this first checks if name is defined in hello function, then checks in greet function and there it finds it so it stops

    hello()

greet()
```

```python
x = 50

def func(x):
    print(f'X is {x}')
    
    #Local 
    x  = 200 ##scope only local to function 
    print(f'I just locally changed X to {x}')
```

```python
func(x)

print(x)
```

```python
x = 50

def func():
    global x
    print(f'X is {x}')
    
    #Local 
    x  = 200 ##this will cahnge the global variable as well
    print(f'I just locally changed X to {x}')
```

```python
print(x) ##variable will be reassigned only when I run the function 
```

```python
func()
```

```python
print(x)
```
