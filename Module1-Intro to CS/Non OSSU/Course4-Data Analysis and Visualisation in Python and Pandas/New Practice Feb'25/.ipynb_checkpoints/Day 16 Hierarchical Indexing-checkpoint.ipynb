<h3> Creating multi index with set index</h3>

```python
pops = pd.read_csv("data/state_pops.csv")
pops.head(1)
```

```python
pops.set_index("state")
```

```python
pops.set_index(["state", "year"], inplace = True)  ## will create multi index
```

```python
pops.head(10)
```

<h3> Sorting a multi index</h3>

```python
pops.index
```

```python
pops.sort_index(inplace = True) ## will sort index at multi level but not in place of course
```

```python
pops.sort_index(ascending = False)
```

```python
pops.sort_index(level = 1) ## this will sort based on level 1 which is year
```

```python
pops.sort_index(ascending = (False, True)) ## this will arrange index in descending for state and asengin for year
```

<h3> Using loc with multiindex</h3>

```python
pops.loc[["CA", "NV"]].head(1)
```

```python
##using multiple levels in loc

pops.loc[[("NV", 1992), ("CA", 1990)]]
```

```python
pops.loc[("AK", 1990):("AK",1995)]
```

```python
pops.loc[:,[1990,1992], :] ## this means all level 0 values are ok and we want that for 1990 year
```

<h3> Cross Section with XS Method </h3>

```python
# Better way of wriging this pops.loc[:,[1990,1992], :] is thorugh the xs or cross section method

pops.xs(2013, level = 1) ## however this can work on only a single vale of the index and is not flexible to cpautre multiple values liek loc
```

<h3> Get level values </h3>

```python
pops.index.levels ## will give level values for both the levels 
```

```python
pops.index.get_level_values(0) ## we specify the level for which we want to get the values
```

```python
even_years = pops.index.get_level_values(1) % 2 == 0 ## this will get a mask of which all years are even
pops[even_years
]
```

```python
##Filtering data for all states which end with A

states_A = pops.index.get_level_values(0).str[1] == "A"
pops[states_A]
```

<h3> Hierarchical columns </h3>

```python
df = titanic.groupby(["pclass","sex"]).mean(numeric_only = True)
```

```python
##lets use this df to analyse 
df
```

```python
df.loc[(2, "male")] ## getting the index of pclass 2 and sub index of male
```

```python
df = titanic.groupby("sex").agg({
    "age": ["min", "max", "mean"],
    "fare": ["min", "max", "mean"],
    "survived" : ["mean"]}
)
```

```python
df ##here index is not multi level but columns are multi levels
```

```python
df.index
```

```python
df.columns
```

```python
##If I have to access a particular multi index column

df[("age", "mean")]
```

```python
## Another way of ding this
df[("age", "mean")]
```

<h3> Stack and unstack </h3>

```python
pops.unstack() ## this will make it  a single level index and add the 2nd leevl to columns making columns multi level
```

```python
unstacked_df = pops.unstack("state") ## will unstack the 1st level
```

```python
unstacked_df.stack() ## this will reverse the operation
```

<h3> Plotting with unstack </h3>

```python
titanic.groupby(["pclass", "sex"])["age"].mean().plot(kind = "bar") ##this will not be so useful
```

```python
## we can unstack it which will be more meaningful

titanic.groupby(["pclass", "sex"])["age"].mean().unstack().plot(kind = "bar")
```

```python
titanic.groupby(["pclass", "sex"])["age"].mean().unstack(level = "pclass").plot(kind = "bar")
```

```python
titanic.groupby(["sex", "survived"])["age"].mean().unstack().plot(kind = "bar")
```
