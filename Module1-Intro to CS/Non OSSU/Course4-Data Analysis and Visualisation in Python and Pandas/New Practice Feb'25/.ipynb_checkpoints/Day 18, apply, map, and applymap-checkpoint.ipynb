```python
import pandas as pd
stocks = pd.read_csv("data/car_stocks.csv")
titanic =  pd.read_csv("data/titanic.csv")
titanic["age"]  = pd.to_numeric(titanic["age"], errors = "coerce")
titanic["fare"]  = pd.to_numeric(titanic["fare"], errors = "coerce")
```

<h3> Applying functions to a series </h3>

```python
def years_to_days(yrs):
    return yrs*365

titanic.age.apply(years_to_days) ##this will pass each element through the smile function
```

```python
def age_group(age):
    if age < 2:
        return "infant"
    elif age < 12:
        return "child"
    elif age < 19:
        return "teen"
    elif age < 50:
        return "adult"
    else:
        return "senior"

titanic["age_group"] = titanic.age.apply(age_group)
```

```python
titanic.age_group.value_counts()
```

```python
titanic.groupby(["age_group", "sex"]).survived.mean()
```

<h3> Apply with lambdas and arguments </h3>

```python
titanic["fare"].apply(lambda x: f"${x*24}")
```

```python
## Lets say we want to pass a function with more than one arguments/params

def convert_currency(num, multiplier):
    return f"${num*multiplier}"
```

```python
titanic["fare"].apply(convert_currency, args = (24,)) ## using only convert_currency will throw an error because it expects 2 arguments hence we wass args separately in a tuple
```

<h3> Apply with lambdas and arguments </h3>

```python
df = titanic[["pclass", "survived", "age"]]
```

```python
def get_range(s): ## takes s as an argument
    return s.max() - s.min()

df.apply(get_range) ## will apply the get range function on each of the columns by default

## this will not run on each element of the series, but on overall series itself
```

```python
## applying a function on rows

df.apply(get_range, axis = 1) ##will get the maximun value from any of the column minus the minimu, value from any columns
```

```python
def get_fam_size(s):
    fam_size = s.sibsp + s.parch
    if fam_size == 0:
        return "solo"
    elif fam_size < 5:
        return "average"
    else:
        return "large"

titanic["fam_size"] =titanic.apply(get_fam_size, axis = 1)
```

```python
titanic.fam_size.value_counts()
```

```python
titanic.groupby("fam_size").survived.mean()
```

<h3> Series Map Method </h3>

```python
## can be used to create a new series by mapping value onto another

titanic["pclass"].map({1: "1st", 2:"2md", 3: "3rd"})
```

```python
titanic.age.map(lambda x: x < 18)
```

```python
titanic.age.apply(lambda x: x < 18)
```

<h3> ApplyMap method</h3>

```python
titanic[["name", "sex", "age_group"]].applymap(str.upper)
```
