```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
```

```python
penguins = pd.read_csv("seaborn/penguins.csv")
tips = pd.read_csv("seaborn/tips.csv")
flights = pd.read_csv("seaborn/flights.csv")
```

```python
sns.set_theme() ##to set the default theme of seaborn
sns.scatterplot(data = tips, x = "total_bill" , y = "tip")
```

```python
##we can get the distribution across a particular category very very easily

sns.scatterplot(data = tips, x = "total_bill" , y = "tip", hue = "sex")
```

```python
## Can add one more cut here using style param

sns.scatterplot(data = tips, x = "total_bill" , y = "tip", hue = "sex" , style = "time")
```

```python
## we ca add one more filter on the graph to capture size

sns.scatterplot(data = tips, x = "total_bill" , y = "tip", hue = "sex" , style = "time", size = "size")
```

<h3> Relplot method </h3>

- **This function/method will create multiple plots with a single line of code**
- **scatterplot is an axis level method, while relpot is figure level moethod**


```python
sns.relplot(data = tips, x = "total_bill", y ="tip", kind = "scatter", col = "sex") ## column = sex will create two scatter plaots
```

```python
sns.relplot(data = tips, x = "total_bill", y ="tip", kind = "scatter", hue = "smoker", col = "sex") ## column = sex will create two scatter plaots
```

```python
sns.relplot(data = tips, x = "total_bill", y ="tip", kind = "scatter", hue = "smoker", col = "sex", row = "time")
```

```python
sns.relplot(data = trips, x = "hour", y = "total", kind = "line", col = "pickup_borough", row = "dropoff_borough", hue = "payment")
```

<h3> Resizing seaborn plots </h3>

**plt.figure - to be used for axes level methods but not for figure level methods such as replot**

```python
plt.figure(figsize = (10,5)) 
sns.lineplot(data = flights, x = "year", y = "passengers", hue = "month")
```

**for replot or figure level method figsize wont work**

```python
plt.figure(figsize = (1,5)) ## this ont do anything  , because replot is wokring on its on figure and not this figure
sns.relplot(
    data = tips, 
    x = "total_bill", y ="tip", 
    kind = "scatter", 
    hue = "smoker", 
    col = "sex",
    height = 2, ##this will make the height of the individual facet or the plot to be 2 inches
    aspect = 2 ## this will control the width of the plot , aspect of 2 would mean that the width is twice as large as the height
) ## column = sex will create two scatter plaots
```

<h3> Strip and swarm plots </h3>

**Works with categoricla data such sex, or pickup_borough vs numerical data etc**

```python
plt.figure(figsize = (12,5))
trips_sample = trips.nlargest(600, "total")
sns.swarmplot(data = trips_sample, x = "pickup_borough", y = "distance", hue = "pickup_borough")
```

<h1> Controlling seaborn aestheitics</h1>

5 themes in seaborn only
- white
- dark
- whitegrid
- darkgrid
- ticks

```python
sns.set_style("darkgrid")

sns.scatterplot(data = tips, x = "total_bill", y = "tip")
```

**can also use set_theme method and pass style as param**

```python
sns.set_theme(style = "white")
sns.scatterplot(data = tips, x = "total_bill", y = "tip")
```

```python
sns.set_style({'axes.facecolor': '#CAD3C8',  'grid.color': '#2C3A47',})

sns.lineplot(data = tips, x = "total_bill", y = "tip")
plt.grid(True)
```

<h3> Despine method </h3>

```python
sns.set_style("white")
sns.scatterplot(data = tips, x = "total_bill", y= "tip")
```

**Spine refers to the 4 borders on top , ledt right, bottom**

```python
sns.scatterplot(data = tips, x = "total_bill", y= "tip")
sns.despine() ##by default , it woll remove top and right border
```

```python
## can also specify left and bottom borders 

sns.scatterplot(data = tips, x = "total_bill", y= "tip")
sns.despine(left=  True, bottom = True)
```

<h3> Changing color palettes </h3>

Doesnt dictate the background or the edge color etc but rather the fill color etc

- Deep
- Muted
- Bright
- Pastel
- Dark
- Colorblind

```python
sns.barplot(data = tips, x = "day", y = "total_bill", hue = "day")
```

```python
sns.set_palette("pastel")
sns.barplot(data = tips, x = "day", y = "total_bill", hue = "day")
```

```python
sns.color_palette("ocean", 20) ## will give 5 colors from palette ocean
```

```python
sns.color_palette("gist_rainbow", 20)
```

```python
sns.set_palette("terrain")
sns.barplot(data = tips, x = "day", y = "total_bill", hue = "day")
```

```python
sns.barplot(data = tips, x = "day", y = "total_bill", hue = "day" , palette = "Blues") ## can define palette inside a code keeping the original global palette consistent
```
