<h3> Stripping whitespace with strip</h3>

```python
s = pd.Series(['1. Hawk.  ', '2. Pickle!\n', '3. Melonhead?\t'])
```

```python
s
```

```python
## Series has trailing spaces and characters which can be managed by strip function
```

```python
s.str.strip() ## wil remove trailining spaces
```

```python
##Can specifty which string we want to remove either leading or trailining
s.str.strip(to_strip = "123. \n \t") ## will only remove leading or trailning characters not from the middle of the string
```

```python
# we can use lstrip or rstrip to remove leading or trailing characters specifically

s.str.rstrip()
```

```python
s.str.lstrip(to_strip = "123.")
```

<h3> Splitting text values with Split</h3>

```python
titanic["home.dest"].str.split() ## will split by default on space characters and will create a list of the divided strings
```

```python
##we can specificy byw whcih element we want to split

titanic["home.dest"].str.split("/")
```

```python
## we can create new colmns with the split values

titanic["home.dest"].str.split("/", expand = True )
```

```python
titanic["home"] = titanic["home.dest"].str.split("/", expand = True )[0]
titanic["destination"] = titanic["home.dest"].str.split("/", expand = True )[0] ## adding new columns to original df
```

```python
titanic.head(1)
```

```python
## can even define how many times we want to split on

titanic["home.dest"].str.split("/", expand = True , n = 1)
```


## Replacing portions of result with replace

```python
ufo = pd.read_csv("data/nuforc_reports.csv")
```

```python
ufo.duration.replace("seconds", "s") ##important to note that this will work on entire strings so will look for seconds on its own and should not be a smaller part of any larger string
```

```python
ufo.duration.str.replace("seconds", "s") ## this will replace seconds in form of a substring as well
```

```python
ufo.duration.str.replace("seconds|minutes", "", regex = True) ## this will replace either seconds or minutes by a blank
```

```python
## Can even use a clalable inside a replace method

def abbr(reo): ##reo refers to a regex object
    return reo.group()[0]
```

```python
ufo.duration.str.replace("seconds|minutes|hours", abbr, regex = True)
```

<h3> Testing string with contains</h3>

```python
ufo[ufo["duration"].str.contains("hour", na = False)] ## will return all rows where duration has the string hour in it 
```
