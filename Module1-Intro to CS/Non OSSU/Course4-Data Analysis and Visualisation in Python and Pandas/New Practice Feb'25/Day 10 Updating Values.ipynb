<h3> Updating multiple values using loc</h3>

```python
## we can add a new column even if that column does not exist
countries.loc[["Denmark", "Norway", "Sweden", "Iceland", "Finland"], ["Regional indicator" , "IsCold"]] = ["Nordic", "Yes"]
countries.loc[["Denmark", "Norway", "Sweden", "Iceland", "Finland"]]

## in this example regional indicator values are being replaced by Nordic and new columns is being crated iscold where value = Yes for only these 5 countries
```

<h3> Making updates with loc and boolean masks </h3>

```python
many_bedrooms = houses["bedrooms"] >= 10
houses.loc[many_bedrooms, ["bedrooms"]] = "Too many"
```

```python
houses[houses["bedrooms"] == "Too many"]
```

```python
high_quality = houses["grade"] > 12
great_views = houses["view"] == 4

houses.loc[great_views & high_quality, ["luxury_level"]] =  "ultra"
```

```python
houses.luxury_level.value_counts()
```
