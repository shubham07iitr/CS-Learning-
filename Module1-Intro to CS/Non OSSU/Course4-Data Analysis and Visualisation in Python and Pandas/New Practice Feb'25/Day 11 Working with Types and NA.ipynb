<h3> Introducing the category type </h3>


**Category datatype would mean the variable can take only a finit no of values such as M or F**

Where this would help is if we have a column with Object values it can take any value so it will set aside or use much more space vs category

In the above case before casting sex as category dtype we had 154kb of memory usage, after it should reduce

```python
titanic["sex"] = titanic.sex.astype("category")
titanic.info()
```

**as you see the memory usage has reduced to 144.7kb**

This is because if we use object dtype, even though we are using same two values, but memory has to be allotted for each item, however if it is category memory has to be allotted for only 2 items and each row just stores the reference to the memory of the two elements

```python
titanic["sex"] = "MALE" ##If I run this it will go back to being an object
```

```python
titanic.info()
```

<h3> dropna</h3>

```python
##dropna will give an updated series dropping the values of Na

stats["assists"].dropna()
```

```python
stats.dropna() ##for df - it will drop any row which have even a single NaN value is there
```

```python
stats.dropna(how = "all") ## this will drop only those rows where all values are Na
```

```python
## focus only on a particular column/s with Na values

stats.dropna(subset= "league") ## this will drop Na values for any rows where league is Na
```

```python
stats.dropna(axis = 1) ## this will drop any column which has any Na value
```
