<h3> Combining conditions using AND and OR </h3>

```python
women = titanic["sex"] == "female"
died = titanic.survived == 0 
titanic[women & died].head(1)
```

```python
water = (houses["waterfront"] == 1)
price = (houses["price"] < 500000)
houses[water & price].head(5)
```

```python
houses[(houses["view"] ==  4) & (houses["grade"] >= 10)].head(5)
```

```python
##Using the OR condition using the '|' separator

houses[((houses["yr_built"] >= 2010) | (houses["yr_renovated"] >= 2010)) & houses.bedrooms >=  1]
```

```python
nf[nf["director"].isin(["Martin Scorcese", "David Fincher"])].head(5)
```

<h3> bitwise negation </h3>

```python
df = titanic.head()
women = df.sex == 'female'
~women ##this will give all cases where this is not true
```

```python
newly_built = houses["yr_built"] >= 2010
newly_renovated = houses["yr_renovated"] >= 2010
recent_homes = newly_built | newly_renovated
houses[~recent_homes].head(5)
```

<h3> isna and notna </h3>

```python
nf[nf["director"].isna()].head(5)
```

```python
nf[nf["director"].isna() & nf["cast"].isna()].head(5)
```

```python
nf[nf["director"].notna() & nf["cast"].isna()].head(5)
```
