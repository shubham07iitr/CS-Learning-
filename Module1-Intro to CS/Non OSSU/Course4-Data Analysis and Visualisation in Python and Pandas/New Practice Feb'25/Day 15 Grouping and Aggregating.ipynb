```python
titanic =  pd.read_csv("data/titanic.csv")
titanic["age"]  = pd.to_numeric(titanic["age"], errors = "coerce")
titanic.info()
```

```python
df.head(1)
```

```python
gbo = df.groupby("sex") ##this will just create a groupby object and not actually return anything
```

```python
gbo.ngroups
```

```python
gbo.first() ## will give the first member of each element of the group
```

<h3> Split, Apply, Combine </h3>

- **Split the data into different groups based on some criteria**
- **Apply function(s) to each group indpeendently**
- **Combine the results into a new data structure**

```python
gbo["age"].mean() ##will give avg age of differnet groups
```

```python
titanic.groupby("pclass")["age"].mean()
```

```python
titanic.groupby("sex")["pclass"].mean()
```

```python
titanic.groupby("sex").mean(numeric_only = True) ## will give mean for each group for each of the remaining columns
```

<h3> Using the agg function </h3>

**running multiple functions on a groupby object**

```python
titanic.groupby("sex")["age"].agg(["min", "max", "mean", "median"]) ## mention all the functinos you want the operation to run on 
```

**Can even specify that for which column we want to run which type of functions**

```python
titanic.groupby("sex").agg({"age": ["min", "max"], 
                                   "pclass": ["mean", "median"]})
```

<h3> Agg with custom functions </h3>

```python
def range(x):
    return x.max() - x.min()

titanic.groupby("pclass")["age"].agg(range) ## agg function automatically passes x param to range function and we dont have to specifically pass it 
```

```python
def count_nulls(series):
    return series.size - series.count()

titanic.groupby("pclass")["age"].agg(count_nulls)
```
