<h3> loc and iloc</h3>

Full Forms of loc and iloc in Pandas:
loc  "Label-based Location"

Used for label-based indexing, meaning you select rows and columns using labels (index names or column names).
Example:
df.loc[2, "price"]  # Selects value at index 2, column "price"

iloc  "Integer Location"
Used for integer-based indexing, meaning you select rows and columns using integer positions (like a list).
Example:
df.iloc[2, 1]  # Selects value at row index 2, column index 1
