<h3> Anatomy of Plots</h3>

```python
import numpy as np
nums= np.arange(5) ##will create an array with 0 to 5 evenly distributed
nums[0:4]
```

```python
plt.plot(nums, nums)
plt.plot(nums, nums*nums)
plt.plot(nums, nums*nums*nums) ## all 3 separate graphs appear on the same canvas
```

```python
##to plot separately on separate figures we have to use plt.figure() 

plt.figure()
plt.plot(nums, nums)
plt.figure()
plt.plot(nums, nums*nums)
plt.figure()
plt.plot(nums, nums*nums*nums) ## all 3 separate graphs appear on the same canvas
```

<h3> Creating bar plots </h3>

```python
plants = ["spinach", "turnip", "rhubharb" ,"broccoli", "kale"]
died = [10,25,5, 30,21]
germinated = [74, 88, 56, 69, 59]
```

```python
plt.figure(figsize = (4,4))
plt.bar(plants, germinated, color = "red", width = 0.5)
plt.title("Germination rates")
plt.ylabel("No of seedlings")
```

```python
plt.figure(figsize = (4,3))
plt.bar(plants, germinated, color = "red", width = 0.5, label = "germinated")
plt.bar(plants, died, color = "olive", width = 0.5, label = "died" , bottom = germinated) ## this will ensure that died is stacked on top of germinated
plt.legend(shadow = True, frameon = True, facecolor = "white")
plt.show()
```

```python
## can even run a horizontal bar plot using barh 
```

```python
plt.figure(figsize = (4,4))
plt.barh(plants, germinated, color = "red", height = 0.5)
plt.title("Germination rates")
plt.ylabel("No of seedlings")
```

<h3> Subplots </h3>

```python
nums = np.arange(5)
```

```python
plt.figure(figsize = (10,4))
plt.suptitle("Subplot", fontsize = 30)
plt.subplot(1,3,1) ##this means we have 1 row, 3 columns, and 1st index
plt.plot(nums, nums)
plt.title("X")
plt.subplot(1,3,2)
plt.title("X squared")
plt.plot(nums, nums**2, color = "red")
plt.subplot(1,3,3)
plt.title("X cubed")
plt.plot(nums, nums**3, color = "yellow")
plt.tight_layout()
```

```python
plt.figure(figsize = (8,8))
plt.subplot(2,2,1) # I want 2 rows 2 colums and working on 2st plot
plt.plot(nums, nums, color = "yellow")
plt.subplot(2,2,2)
plt.plot(nums, nums, color = "red")
plt.subplot(2,2,3)
plt.plot(nums, nums, color = "blue")
plt.subplot(2,2,4)
plt.plot(nums, nums, color = "green")
```

<h3> Putting it all together </h3>

```python
plt.style.use("fivethirtyeight")
first_class = titanic[titanic.pclass == 1]["age"]
second_class = titanic[titanic.pclass == 2]["age"]
third_class = titanic[titanic.pclass == 3]["age"]

plt.figure(figsize = (10,4))
ax = plt.subplot(1,3,1) ##will return an axes object
plt.hist(first_class, color = "blue", label = "1st class", alpha = 0.5) ##alpha will give us transparency
plt.title("1st class")
plt.legend()
plt.subplot(1,3,2, sharey = ax)
plt.hist(second_class, color = "red", label = "2nd class", alpha = 0.5)
plt.title("2nd class")
plt.legend()
plt.subplot(1,3,3, sharey = ax)
plt.hist(third_class, color = "green", label = "3rd class", alpha = 0.5)
plt.title("3rd class")
plt.legend()
plt.show()
```
