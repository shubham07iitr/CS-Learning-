##### Have you noticed that the type of Year is int64. But pandas has a different type to work with Time Series. Let's see it now.

### Step 5. Convert the type of the column Year to datetime64

```python
crime["Year"] = pd.to_datetime(crime["Year"].astype(str), format = '%Y').dt.year
crime.head(2)
```

### Step 6. Set the Year column as the index of the dataframe

```python
crime.set_index("Year").head(1)
```

### Step 7. Delete the Total column

```python
crime.drop(columns = ["Total"])
crime.head(1)
```

### Step 8. Group the year by decades and sum the values

#### Pay attention to the Population column number, summing this column is a mistake

```python
def decade(year):
    if year >= 1960 and year <1970:
        return "60s"
    elif year >= 1970 and year <1980:
        return "70s"
    elif year >= 1980 and year <1990:
        return "80s"
    elif year >= 1990 and year <2000:
        return "90s"
    elif year >= 2000 and year <2010:
        return "2000s"
    else:
        return "2010s"

crime["decade"] = crime.Year.apply(decade)
crime.head()
crime.groupby("decade").mean()
```

### Step 9. What is the most dangerous decade to live in the US?

```python
crime.groupby("decade")["Total"].sum().sort_values(ascending = False)
```
