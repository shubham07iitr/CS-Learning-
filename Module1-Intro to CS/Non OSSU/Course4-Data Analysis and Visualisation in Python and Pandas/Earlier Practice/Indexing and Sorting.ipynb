```python
houses = pd.read_csv("DataAnalysis/Data/kc_house_data.csv")
titanic = pd.read_csv("DataAnalysis/Data/titanic.csv")
```

```python
houses.sort_values("price", ascending= False)
```

<h2> Can sort based on multiple columns

```python
houses.sort_values(["bedrooms", "bathrooms"], ascending = False)
```

```python
titanic.sort_values("name").head(5) ##uppercase comes before lower case like ABC....Zabc...z
```

```python
titanic.sort_values("name", key = lambda col:col.str.lower()) ##this will ensure that if small cap is there but a then its still at top
```

```python
countries.sort_index().head(2) #does not work inplace
```

```python
titanic.pclass.value_counts().sort_values().plot(kind = "bar")
```
