<h4> Saving for a house </h4>

```python
yearly_salary = float(input("Please enter your yearly salary"))
portion_saved = float(input("What % you want to save every month"))
cost_of_dream_home = float(input("Cost of your dream home"))

portion_down_payment = 0.25
amount_saved = 0
rate_of_return = 0.05
no_of_months = 0
down_payment_required = portion_down_payment*cost_of_dream_home

while amount_saved < down_payment_required:
    amount_saved = amount_saved + amount_saved*(rate_of_return/12) + portion_saved*(yearly_salary/12)
    no_of_months += 1

print(f"Number of months required to amass the down payment is {no_of_months}")
```

<h4> Saving with a raise</h4>

```python
yearly_salary = float(input("Please enter your yearly salary"))
portion_saved = float(input("What % you want to save every month"))
cost_of_dream_home = float(input("Cost of your dream home"))
semi_annual_raise = float(input("Enter semi-annual raise as a decimal"))

portion_down_payment = 0.25
amount_saved = 0
rate_of_return = 0.05
no_of_months = 0
down_payment_required = portion_down_payment*cost_of_dream_home

while amount_saved < down_payment_required:
    amount_saved = amount_saved + amount_saved*(rate_of_return/12) + portion_saved*(yearly_salary/12)
    no_of_months += 1
    if no_of_months % 6 == 0:
        yearly_salary = yearly_salary*(1+semi_annual_raise)

print(f"Number of months required to amass the down payment is {no_of_months}")
```

<h4> Chosing an interest rate </h4>

```python
initial_deposit = float(input("Enter the initial deposit:"))
cost_of_dream_home = 800000
portion_down_payment =  0.25
down_payment_required = cost_of_dream_home * portion_down_payment
r_lower = 0
r_upper = 1
steps = 0

while True:
    if initial_deposit*(1+ r_upper/12)**36 < down_payment_required - 100:
        r_binary = None 
        break
    elif initial_deposit > down_payment_required - 100:
        r_binary = 0.0
        break
    r_binary = (r_lower + r_upper)/2
    amount_saved = initial_deposit*(1+r_binary/12)**36
    if amount_saved >= down_payment_required - 100 and amount_saved <= down_payment_required + 100:
        break
    elif amount_saved >= down_payment_required + 100:
        r_upper = r_binary
    else:
        r_lower = r_binary
    steps += 1

print(f"Min rate of return required is {r_binary}")
print(f"Steps in the binary search are {steps}")
        
        
    
```
